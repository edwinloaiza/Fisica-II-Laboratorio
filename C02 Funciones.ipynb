La teoría electromagnética es la rama de la Ciencia Física que se ocupa de los campos eléctricos y magnéticos, así como de sus interacciones entre sí y con objetos físicos. 

Por lo general, se piensa en la electricidad como el flujo de electrones en un conductor, o como electricidad estática, o relámpagos. El magnetismo se asocia con imanes permanentes, el magnetismo terrestre, agujas de brújula, electroimanes, patrones en limaduras de hierro, entre otros.  

Los elementos físicos, como electrones y partículas cargadas por un lado, y materiales magnéticos por el otro, son solo una parte del fenómeno. Más poderosas son las fuerzas invisibles que atraen o repelen objetos, permitiendo máquinas eléctricas, motores y generadores transformar la energía en formas útiles. También posibilitan la transmisión inalámbrica de información y, por ende, de energía a distancias grandes o pequeñas. 

Aunque no se aborde aquí, la información es una forma de energía, estrechamente ligada a todos los fenómenos electromagnéticos. Los campos electromagnéticos no existen sin manifestaciones físicas, y las manifestaciones físicas no existen sin los campos invisibles. 

Para comprender estas complejas interacciones, se tuvo que desarrollar una forma especial de matemáticas avanzadas, más compleja que el cálculo utilizado para determinar el movimiento de los cuerpos físicos. Esta rama de las matemáticas abarca la energía difusa distribuida en el espacio y dentro de áreas confinadas. 

Su primer desarrollo se produjo con el estudio de la distribución del calor dentro de un sólido, el flujo de fluidos en tuberías y el vapor en todas sus manifestaciones. Cuando se descubrió que la electricidad se comportaba de manera muy similar a un fluido, las matemáticas se aplicaron entonces a las corrientes eléctricas. La distribución de calor dentro de un sólido se convirtió en la distribución de campos eléctricos y magnéticos, así como en variaciones en la densidad de corriente.


Las matemáticas y el simbolismo utilizados en electromagnetismo suelen resultar desconocidos e imponentes para quienes no están familiarizados con ellos. Con ecuaciones diferenciales parciales extensas, campos y vectores difíciles de visualizar, y símbolos inusuales como $\nabla$, $\nabla^2$, $\nabla\cdot$, $\nabla\times$, $\partial$,$\oint$,$\oiint$ ,,$\oiiint$,... , etc., el tema suele ser espinoso y difícil de abordar.

## Funciones

Consideraremos funciones con dominio en un espacio $n-$ dimensional  $\mathbb{R}^n$ y codominio o rango en un espacio $m-$ dimensioal  $\mathbb{R}^m$.
\begin{align}
\mathbf{F}:\mathbb{R}^n\to \mathbb{R}^m
\end{align}
- Cuando $n=m=1$, tales funciones son llamadas **funciones reales de variable real**.
- Cuando $n=1$ y $m>1$ es un a **función vectorial de variable real**.
- Cuando  $n>1$ y $m=1$ es una función real de variable vectorial o **campo escalar**.
- Cuando $n>1$ y $m>1$ es una función vectorial de variable vectorial o simplemente **campo vectorial**.

### Funciones reales de variable real

Las funciones reales de valor real:
\begin{equation}
F:\mathbb{R}\to \mathbb{R}
\end{equation}
Fueron estudiadas principalmente en el curso de Física I en el tema de cinemática, cuando se describe la posición $x$ de una partícula en función del tiempo $t$, que denotamos como $x=x(t)$, la rapidez y la acelearación se definen a partir de la primer y segunda derivada respectivamente.
\begin{align}
v(t)&=\frac{d\,x}{dt}\\
a(t)&=\frac{d\,v}{dt}=\frac{d^2\,x}{dt^2}\\
\end{align}
En el caso de un movimiento de caída sobre la superficie de la Tierra, la posición $y$ en función del tiempo $t$ está dada por
$$
y(t)=-\frac{g}{2}t^2+v_{y\,0}\,t+y0
$$
donde $y_0$ es la posición inicial, $v_{y\,0}$ es la la rapidez inicial y $g=9.8\,m/s^2$ es la aceleración de la gravedad.


##### Ejemplo: Caída libre


En el caso de un objeto que se lanza hacia arriba desde la terraza de un edificio de altura $h=20\,m$ con rapidez inicial $v_{y\,0}=20\,m/s$, se tiene que:

\begin{align}
y(t)&=20+20\,t-\frac{g}{2}t^2\\
v(t)&=20-g\,t\\
a(t)&=-g\\
\end{align}

```python
#@ Gráficas de posición, velociad y aceleración de un cuerpo que semueve verticalmente sobre la superficie de la Tierra.
# Importar librerías
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
t = sp.Symbol('t')  # Se define la variable simbólica t como el tiempo
g = 9.8   # Aceleración debida a la gravedad (m/s^2)
v0 = 20   # Rapidez inicial (m/s)
y0 =  10  # Posición inicial en y

# Definir las funciones simbólicas para y(t), v(t) y a(t)
yt = y0 + v0 * t - 0.5 * g * t**2
vt = yt.diff(t)
at = vt.diff(t)

# Ecuación de tiempo de vuelo (componente y igualada a cero)
tv = sp.solve(yt, t)

# Crear un rango de 100 valores de t para la representación gráfica
t_val = np.linspace(0, float(tv[1]), 100)

# Evaluar las funciones en el rango de tiempo
yt_val = np.array([yt.subs(t, val) for val in t_val])
vt_val = np.array([vt.subs(t, val) for val in t_val])
at_val = np.array([at.subs(t, val) for val in t_val])

# Crear subgráficos
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# Graficar posición vs tiempo
axs[0].plot(t_val, yt_val, label='Posición  $y(t)$')
axs[0].set_xlabel('Tiempo $t(s)$')
axs[0].set_ylabel('Posición  $y(m)$')
axs[0].set_title('Posición $y$ en función del Tiempo $t$')
axs[0].legend()
axs[0].grid(True)

# Graficar velocidad vs tiempo
axs[1].plot(t_val, vt_val, label='Velocidad  $v(t)$',color='green')
axs[1].set_xlabel('Tiempo $t(s)$')
axs[1].set_ylabel('Velocidad $v(m/s)$')
axs[1].set_title('Velocidad $v$ en función del Tiempo $t$')
axs[1].legend()
axs[1].grid(True)

# Graficar aceleración vs tiempo
axs[2].plot(t_val, at_val, label='Aceleración  $a(t)$',color='red')
axs[2].set_xlabel('Tiempo $t(s)$')
axs[2].set_ylabel('Aceleración  $a(m/s^2)$')
axs[2].set_title('Aceleración $a$ en función del Tiempo $t$')
axs[2].legend()
axs[2].grid(True)

# Ajustar el diseño
plt.tight_layout()
plt.show()
```

##### Ejemplo: Movimiento armónico simple

Para un sistema masa-resorte, se considera una masa $m=1\,kg$ unida a un resorte ideal con una constante elástica $k=10\,n/m$. Inicialmente, la masa se encuentra en reposo ($v(0)=0$) y el resorte está estirado a una distancia $y_0=1\,m$.  Se busca graficar las funciones de posición ($y$), velocidad ($v$), y aceleración ($a$) en función del tiempo.

Se puede utilizar la ecuación del movimiento armónico simple (MAS) para describir el comportamiento del sistema masa-resorte:
$$
y(t)=y_0\,\cos(\omega\,t)
$$
Donde $\omega=\sqrt{\frac{k}{m}}$ es la frecuencia angular.

```python
# Importar librerías
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
t = sp.Symbol('t')  # Se define la variable simbólica t como el tiempo
k = 10   # Aceleración debida a la gravedad (m/s^2)
m = 1    # Masa
v0 = 0   # Rapidez inicial (m/s)
y0 =  1  # Posición inicial en y
omega = np.sqrt(k/m)
P = 2*np.pi/omega

# Definir las funciones simbólicas para y(t), v(t) y a(t)
yt = y0 * sp.cos(omega * t)
vt = yt.diff(t)
at = vt.diff(t)

# Calcular el tiempo de vuelo
tv = 2 * np.pi / omega

# Crear un rango de 100 valores de t para la representación gráfica
t_val = np.linspace(0, float(tv), 100)

# Evaluar las funciones en el rango de tiempo
yt_val = np.array([yt.subs(t, val) for val in t_val])
vt_val = np.array([vt.subs(t, val) for val in t_val])
at_val = np.array([at.subs(t, val) for val in t_val])

# Crear subgráficos
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# Graficar posición vs tiempo
axs[0].plot(t_val, yt_val, label='Posición  $y(t)$')
axs[0].set_xlabel('Tiempo $t(s)$')
axs[0].set_ylabel('Posición  $y(m)$')
axs[0].set_title('Posición $y$ en función del Tiempo $t$')
axs[0].legend()
axs[0].grid(True)

# Graficar velocidad vs tiempo
axs[1].plot(t_val, vt_val, label='Velocidad  $v(t)$', color='green')
axs[1].set_xlabel('Tiempo $t(s)$')
axs[1].set_ylabel('Velocidad $v(m/s)$')
axs[1].set_title('Velocidad $v$ en función del Tiempo $t$')
axs[1].legend()
axs[1].grid(True)

# Graficar aceleración vs tiempo
axs[2].plot(t_val, at_val, label='Aceleración  $a(t)$', color='red')
axs[2].set_xlabel('Tiempo $t(s)$')
axs[2].set_ylabel('Aceleración  $a(m/s^2)$')
axs[2].set_title('Aceleración $a$ en función del Tiempo $t$')
axs[2].legend()
axs[2].grid(True)

# Ajustar el diseño
plt.tight_layout()
plt.show()
```

#### Ejemplo: Tiro parabólico

En el cado de un tiro parabólico, la ecuaciones paramétricas de movimiento para los ejes $x$, $y$ son:
\begin{align}
x(t)&=v_{0}\,\cos(\theta)\,t&y(t)&=y_0+v_{0}\,\sin(\theta)\,t-\frac{g}{2}t^2 \\
v_x(t)&=v_{0}\,\cos(\theta)&v_y(t)&=v_{0}\,\sin(\theta)-g\,t \\
a_{x}(t)&=0&a_y(t)&=-g\\
\end{align}
Las funciones que describen el movimiento alo largo de cada eje son funciones reales de variable real.

Despejando $t$ de la ecuación $x(t)$ y rremplazandolo en $y(t)$ se elimina el parámetro tiempo y se obtiene la ecuación cartesiana de la trayectoria:
\begin{align}
y(x)&=y_0+\tan(\theta)\,x-\frac{g}{2\,v_0^2\,\cos^2(\theta)}x^2\\
\end{align}
La ecuación resultante es una función real de variable real que describe una trayectoria parabólica.


```python
from sympy import Symbol, cos, pi, tan
import sympy as sp

# Define the function y(x)
def y(x, v0=20, y0=20, theta=pi/6, g=9.8):
    """Calculates the y-coordinate of the projectile at a given x-coordinate.

    Args:
        x: Symbolic variable representing the horizontal distance.
        v0: Initial velocity.
        y0: Initial height.
        theta: Launch angle.
        g: Acceleration due to gravity.

    Returns:
        The y-coordinate of the projectile.
    """
    return y0 + tan(theta) * x - g * x**2 / (2 * v0**2 * cos(theta)**2)
    
# Create the symbolic variable x
x = Symbol('x')

# Set up the equation y(R) = 0
equation_y_zero = y(x) - 0

# Try to solve for R symbolically using solve with simplify=True
solutions_R = sp.solve(equation_y_zero, x, simplify=True)

# If a solution is found, extract the first positive solution (assuming upward launch)
if solutions_R:
    for sol in solutions_R:
        if sol.evalf() > 0:  # Ensure positive solution for distance
            R = sol
            break  # Take only the first valid solution



# Define the derivative of y(x)
dy_dx = sp.diff(y(x), x)

# Set the derivative equal to zero and solve for x
equation_dy_dx_zero = dy_dx - 0

# Solve for R symbolically using solve
solutions_X = sp.solve(equation_dy_dx_zero, x, simplify=True)

# If a solution is found, extract the first positive solution (assuming upward launch)
if solutions_X:
    for sol in solutions_X:
        if sol.evalf() > 0:  # Ensure positive solution for distance
            X = sol
            break  # Take only the first valid solution

# Calculate the maximum height using the solution
ymax = y(X)


# Plot the function
sp.plot(y(x), (x, 0, R), label='Projectile Trajectory')
plt.show()
```

### Funciones vectoriales

#### Posición en $2\,D$


A partír de las ecuaciones parámetricas del tiro parabólico construimos la ecuación vectorial:
$$
\vec{r}(t)= \left(v_{0}\,\cos(\theta)\,t\right)\mathbf{i}+\left(y_0+v_{0}\,\sin(\theta)\,t-\frac{g}{2}t^2 \right)\mathbf{j}
$$
en este caso a cada instante de tiempo $t_0$ le corresponde un punto $(x_0,y_0)$

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
ti = sp.symbols('ti')  # Se define la variable simbólica ti como tiempo
g = 9.8  # Aceleración debida a la gravedad (m/s^2)
v0 = 20  # Rapidez inicial (m/s)
y0 = 10  # Posición inicial en y
theta = 20 * np.pi / 180  # Ángulo inicial (radianes)

# Definir las funciones simbólicas para r(t) y v(t)
rt = sp.Matrix([v0 * sp.cos(theta) * ti, y0 + v0 * sp.sin(theta) * ti - 0.5 * g * ti**2])
vt = rt.diff(ti)

# Ecuación de tiempo de vuelo (componente y igualada a cero)
equation_y = rt[1]
equation_t = sp.solve(equation_y, ti)

# Obtener el valor positivo de tiempo de vuelo
tv_positive = [sol.evalf() for sol in equation_t if sol.evalf() > 0][0]

# Crear un rango de 10 valores de t para evaluar la velocidad
t_values = np.linspace(0, float(tv_positive), 10)
# Crear un rango de 100 valores de t para evaluar la posición
t_rvalues = np.linspace(0, float(tv_positive), 100)

# Evaluar las funciones simbólicas en los 100 puntos
r_rvalues = np.array([np.array(rt.subs(ti, val)).astype(float) for val in t_rvalues])

# Graficar la función vectorial r(t)
plt.plot(r_rvalues[:, 0], r_rvalues[:, 1], label=r'$\vec{\mathbf{r}}(t) =\langle y_0 + v_0\,\cos(\theta)t, y_0 + v_0\,\sin(\theta)t - \frac{g}{2}t^2\rangle$')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfica de la función vectorial r(t)')
plt.grid(True)

# Ajustar el rango de y para que inicie desde 0
plt.ylim(bottom=0)

# Agregar un vector con cola en el origen y cabeza en t = 1
t_arrow = 1
r_arrow = np.array(rt.subs(ti, t_arrow)).astype(float)
plt.quiver(0, 0, r_arrow[0], r_arrow[1], angles='xy', scale_units='xy', scale=1, color='black', label=r'$\vec{r}(1)$')
plt.scatter(r_arrow[0], r_arrow[1])

plt.legend(loc='lower center')
plt.show()
```

#### Posición y velocidad en $2\,D$



Una alternativa interesante para visualizar el comportamiento de una curva en dos dimensiones consiste en incorporar, además de la propia curva, el conjunto de vectores tangentes a la misma en cada punto. Estos vectores representan las velocidades instantáneas del objeto que se desplaza a lo largo de la curva. Formalmente a cada instante  de tiempo $t$ le correspopnde  un vector posición $\vec{p}(t)$ y un vector veclocidad $\vec{v}(t)=\vec{r}'(t)$. De esta forma, se establece una correspondencia entre el tiempo $t$ y un par ordenado en el espacio de cuatro dimensiones $\mathbb{R}^2\times\mathbb{R}^2$.
\begin{align}
\vec{F}:\mathbb{R}\to \mathbb{R}^2\times\mathbb{R}^2
\end{align}
Esta representación permite observar cómo la posición y la velocidad del objeto se relacionan entre sí a lo largo del tiempo. La dirección del vector tangente en cada punto indica la dirección del movimiento instantáneo, mientras que su magnitud representa la velocidad instantánea.


```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
ti = sp.symbols('ti')  # Se define la variable simbólica ti como tiempo
g = 9.8  # Aceleración debida a la gravedad (m/s^2)
v0 = 20  # Rapidez inicial (m/s)
y0 = 10  # Posición inicial en y
theta = 20 * np.pi / 180  # Ángulo inicial (radianes)

# Definir las funciones simbólicas para r(t) y v(t)
rt = sp.Matrix([v0 * sp.cos(theta) * ti, y0 + v0 * sp.sin(theta) * ti - 0.5 * g * ti**2])
vt = rt.diff(ti)

# Ecuación de tiempo de vuelo (componente y igualada a cero)
equation_y = rt[1]
equation_t = sp.solve(equation_y, ti)

# Obtener el valor positivo de tiempo de vuelo
tv_positive = [sol.evalf() for sol in equation_t if sol.evalf() > 0][0]

# Crear un rango de 10 valores de t para evaluar la velocidad
t_values = np.linspace(0, float(tv_positive), 10)
# Crear un rango de 100 valores de t para evaluar la posición
t_rvalues = np.linspace(0, float(tv_positive), 100)


# Evaluar las funciones simbólicas en los 100 puntos
r_rvalues = np.array([np.array(rt.subs(ti, val)).astype(float) for val in t_rvalues])

# Evaluar las funciones simbólicas solo en esos 10 puntos
r_values = np.array([np.array(rt.subs(ti, val)).astype(float) for val in t_values])
v_values = np.array([np.array(vt.subs(ti, val)).astype(float) for val in t_values])

# Reducir la escala de los vectores velocidad
escala = 0.25  # Puedes ajustar el factor de escala según tus necesidades
v_values_reduced = escala * v_values

# Graficar la función vectorial r(t)
plt.plot(r_rvalues[:, 0], r_rvalues[:, 1], label=r'$\vec{\mathbf{r}}(t) = y_0 + v_0\,\cos(\theta)t, y_0 + v_0\,\sin(\theta)t - \frac{g}{2}t^2$')
plt.quiver(r_values[:, 0], r_values[:, 1], v_values_reduced[:, 0], v_values_reduced[:, 1], angles='xy', scale_units='xy', scale=1, color='lightgreen', label='Vectores velocidad')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfica de la función vectorial r(t) con vectores velocidad')
plt.grid(True)
plt.legend()

# Ajustar el rango de y para que inicie desde 0
plt.ylim(bottom=0)

plt.show()
```

#### Posición, velocidad y aceleración en $2\,D$


Final mente podemo incluir en la gráfica tanto el vector velocidad como el vector aceleración. Se generaliza el caso a una función que asigna a cada instante de tiempo $t$, la posición $\vec{r}$, la velocidad $\vec{v}$ y la aceleración $\vec{a}$. De esta forma, se establece una correspondencia entre el tiempo $t$ y una terna ordenado en el espacio de seis dimensiones $\mathbb{R}^2\times\mathbb{R}^2\times\mathbb{R}^2$.
\begin{align}
\vec{F}:\mathbb{R}\to \mathbb{R}^2\times\mathbb{R}^2\times\mathbb{R}^2
\end{align}


```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
ti = sp.symbols('ti')  # Se define la variable simbólica ti como tiempo
g = 9.8  # Aceleración debida a la gravedad (m/s^2)
v0 = 20  # Rapidez inicial (m/s)
y0 = 10  # Posición inicial en y
theta = 20 * np.pi / 180  # Ángulo inicial (radianes)

# Definir las funciones simbólicas para r(t), v(t) y a(t)
rt = sp.Matrix([v0 * sp.cos(theta) * ti, y0 + v0 * sp.sin(theta) * ti - 0.5 * g * ti**2])
vt = rt.diff(ti)
at = vt.diff(ti)

# Ecuación de tiempo de vuelo (componente y igualada a cero)
equation_y = rt[1]
equation_t = sp.solve(equation_y, ti)

# Obtener el valor positivo de tiempo de vuelo
tv_positive = [sol.evalf() for sol in equation_t if sol.evalf() > 0][0]

# Crear un rango de 10 valores de t para evaluar la velocidad y aceleración
t_values = np.linspace(0, float(tv_positive), 10)
# Crear un rango de 100 valores de t para evaluar la posición
t_rvalues = np.linspace(0, float(tv_positive), 100)

# Evaluar las funciones simbólicas en los 100 puntos
r_rvalues = np.array([np.array(rt.subs(ti, val)).astype(float) for val in t_rvalues])

# Evaluar las funciones simbólicas solo en esos 10 puntos
r_values = np.array([np.array(rt.subs(ti, val)).astype(float) for val in t_values])
v_values = np.array([np.array(vt.subs(ti, val)).astype(float) for val in t_values])
a_values = np.array([np.array(at.subs(ti, val)).astype(float) for val in t_values])

# Reducir la escala de los vectores velocidad y aceleración
escala = 0.25  # Puedes ajustar el factor de escala según tus necesidades
v_values_reduced = escala * v_values
a_values_reduced = escala * a_values

# Graficar la función vectorial r(t) con vectores velocidad y aceleración
plt.plot(r_rvalues[:, 0], r_rvalues[:, 1], label=r'$\vec{\mathbf{r}}(t) = y_0 + v_0\,\cos(\theta)t, y_0 + v_0\,\sin(\theta)t - \frac{g}{2}t^2$')
plt.quiver(r_values[:, 0], r_values[:, 1], v_values_reduced[:, 0], v_values_reduced[:, 1], angles='xy', scale_units='xy', scale=1, color='lightgreen', label=r'$\vec{v}(t)$')
plt.quiver(r_values[:, 0], r_values[:, 1], a_values_reduced[:, 0], a_values_reduced[:, 1], angles='xy', scale_units='xy', scale=1, color='red', label=r'$\vec{a}(t)$')
plt.xlabel('x(m)')
plt.ylabel('y(m)')
plt.title('Posición $r(t)$ con vectores velocidad y aceleración')
plt.grid(True)
plt.legend()

# Ajustar el rango de y para que inicie desde 0
plt.ylim(bottom=0)

plt.show()
```

#### Posición en $3\,D$


En el caso tridimensional asociamos a cada instante de tiempo $t$ un punto en el espacio tridimensional.

en este caso a cada instante de tiempo $t$ le corresponde un punto en el espacio.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Parámetros de la espiral
t = np.linspace(0, 10 * np.pi, 1000)  # Rango de valores para el parámetro t
x = t * np.cos(t)  # Coordenada x

y = t * np.sin(t)  # Coordenada y
z = t  # Coordenada z (altura)

# Crear la figura 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Graficar la espiral
ax.plot(x, y, z, label='Espiral 3D', color='b')

# Personalizar la gráfica
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Espiral en 3D')
ax.legend()

# Mostrar la gráfica
plt.show()
```

#### Tiro parabólico en $3\,D$

En el caso del tiro parabólico tridimensional, las condiciones iniciales se establecen como:
\begin{align*}
\vec{r}_0&=z_0\mathbf{k}\\
\vec{v}_0&=v_0\,\sin(\phi_0)\cos(\theta_0)\mathbf{i}+v_0\,\sin(\phi_0)\sin(\theta_0)\mathbf{j}+v_0\,\cos(\phi_0)\mathbf{k}\\
\vec{a}&=-g\mathbf{k}
\end{align*}
Donde:
- $z_0$ es la altura inicial.
- $v_0$ es la rapidez inicial.
- $\phi_0$ es el ángulo de elevación inicial o ángulo cenital inicial.
- $\theta_0$ es el ángulo azimutal inicial.
- $g$ es la aceleración debida a la gravedad

La ecuación vectorial de movimiento para el proyectil en este caso toma la forma:
$$
\vec{r}(t)=\left(v_0\,\sin(\phi_0)\cos(\theta_0)\,t\right)\mathbf{i}+\left(v_0\,\sin(\phi_0)\sin(\theta_0)\,t\right)\mathbf{j}+\left(z_0+v_0\,\cos(\phi_0)\,t-\frac{g}{2}t^2\right)\mathbf{k}
$$
Al analizar la ecuación, se puede observar que la trayectoria del proyectil sigue siendo parabólica y plana. Esto significa que la trayectoria se desarrolla en el plano formado por los vectores $\vec{v}_0$ y $\vec{a}$

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import sympy as sym

ti=sym.Symbol('ti')

# Condiciones iniciales

z0=20
v0=20
phi0=np.pi/6
theta0=np.pi/4
g=9.8

eq=z0 + v0 * sym.cos(phi0) * ti - 0.5 * g * ti**2
tv = sym.solve(eq, ti)[1]

# Parámetros de la trayectoriua

t = np.linspace(0,  float(tv), 100)  # Rango de valores para el parámetro t

v=np.array([v0*np.sin(phi0)* np.cos(theta0),v0*np.sin(phi0)* np.sin(theta0),v0*np.cos(phi0)])

x = v[0]*t  # Coordenada x
y = v[1]*t  # Coordenada y
z = z0+v[2]*t-1/2*g*t**2  # Coordenada z (altura)

# Crear la figura 3D
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Graficar la parábola
escalav=0.5
ax.plot(x, y, z, label='Espiral 3D', color='b')
ax.quiver(0, 0, z0, escalav*v[0], escalav*v[1], escalav*v[2], color='lightgreen')
ax.quiver(0, 0, z0, 0, 0, -g, color='red')

# Añadir etiqueta en la cabeza del vector v0 y g
ax.text(escalav*v[0], escalav*v[1], escalav*v[2]+z0, r'$\vec{v}_0$', fontsize=12, color='lightgreen',ha='right', va='baseline')
ax.text(0,0,z0-g*1.5 ,r'$\vec{g}$', fontsize=12, color='red',ha='right', va='bottom')


# Personalizar la gráfica
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title(r'Tiro parabólico en $3\,D$')
#ax.legend()

# Mostrar la gráfica
plt.show()
```

#### Curvas de nivel $z=\phi(x,y)$

Sea $\phi$ un campo escalar definido en un subconjunto $S$ de $\mathbb{R}^2$ y sea $c$ un número real dado. Un conjunto de nivel de $\phi$ con respecto a $c$ es el conjunto de todos los puntos $(x,y)$ en $S$ donde la función toma el valor $c$: 
$$
C_c=\{(x,y)\in S\vert \phi(x,y)=c\}
$$ 
Para cada una de las funciones escalares $\phi_1,\,\phi_2,\,\phi_3,\,\phi_4,\,$ definidas en la sección anterior, se procederá a trazar determinar el tipo de curva de nivel y la gráfica correspondientes para diferentes valores de $c$.

##### Paraboloide: $\phi_1(x,y)=1-x^2-y^2$. 


Si $c=1-x^2-y^2$. Al considerar $c = 1 - x^2 - y^2$, se deduce que $x^2 + y^2 = 1 - c$. Por consiguiente, es evidente que $1 - c > 0$, lo cual implica $c \leq 1$. Introduciendo la variable $r^2 = 1 - c$, se establece que las curvas de nivel corresponden a circunferencias con radio $r$ centradas en el origen, expresadas como:
$$
r^2=x^2+y^2
$$
Trazar las curvas de nivel para $c=\{1,\,0.5,\,0,\,-0.5,\cdots,\,-3.5\}$

##### Paraboloide hiperbólico: $\phi_1(x,y)=x^2-y^2$. 

Si $c=x^2-y^2$. Al considerar $c =x^2 - y^2$, se tiene que $c$ puede tomar cualquier valor positivo, negativo o cero. 
- Si $c=0$, se tiene que $\vert x\vert=\vert y\vert$, de donde $y=\pm\,x$ genera dos rectas identidad. 
- Si $0<c=a^2$, se tiene que $a^2=x^2-y^2$, genera hipérbolas  que abren sobre el eje $x$ con vértices en: $x=\pm\,a$ y asíntotas: $y=\pm\,x$
- Si $0>c=-b^2$, se tiene que $-b^2=x^2-y^2$, es decir $b^2=y^2-x^2$  genera hipérbolas  que abren sobre el eje $y$ con vértices en: $y=\pm\,b$ y asíntotas: $y=\pm\,x$

Trazar las curvas de nivel para $c=\{-2.0,\,-1.5,\,-1,\,\cdots,\,2.5,\,3.0\}$

##### Potencial eléctrico: $\phi(x,y)=  \frac{1}{\sqrt{(x + 1)^2 + y^2}} - \frac{1}{\sqrt{(x - 1)^2 + y^2}} + \frac{1}{\sqrt{x^2 + (y - 1)^2}} $

Debido a la presencia de raíces cuadradas y la suma de términos no lineales, la función $\phi(x,y)$ presenta una complejidad considerable. La complejidad de la función dificulta la determinación analítica de las curvas de nivel. En otras palabras, no es sencillo encontrar una expresión matemática que represente las curvas de nivel para cualquier valor de la función.

##### Cono circular recto $\phi(x,y)=\sqrt{x^2+y^2}$

Si $0\geq c=\sqrt{x^2+y^2}$, entonces $c^2=x^2+y^2$ lo que implica que las curvas de nivel corresponden a circunferencias con radio $c$, centradas en el origen. Al comparar estas curvas de nivel con las obtenidas para el paraboloide, se puede establecer un hecho significativo: superficies distintas pueden compartir las mismas curvas de nivel.

##### Proyección de contornos


La técnica de proyectar contornos rellenos en un gráfico tridimensional (3D) permite una mejor visualización de la forma y la variación de una función o superficie tridimensional. Esta técnica involucra los siguientes elementos:
- Superficie 3D: La base del gráfico es un conjunto de datos que representa una superficie tridimensional. Esta superficie puede ser definida por una ecuación matemática o por un conjunto de puntos (x, y, z).

- Contornos: Los contornos son líneas o curvas que conectan puntos de igual valor en la superficie. Un gráfico de contornos 2D representa la superficie visualizada desde arriba.

- Relleno: Se añaden colores en los espacios entre las curvas de contorno, lo que mejora la percepción de elevaciones y depresiones en la superficie.

- Proyección en las paredes: Las curvas de nivel se "proyectan" sobre los planos x-z e y-z del gráfico (las "paredes"), creando una especie de sombras en perspectiva de la superficie.

Ventajas:

- Mejor comprensión: Esta técnica ayuda a entender intuitivamente la forma de una superficie 3D. Las proyecciones de contorno en las paredes proporcionan referencias visuales adicionales a la altura y pendiente de la superficie.
- Análisis de datos: Los contornos rellenos revelan patrones, picos, valles y otras características importantes de una superficie 3D que se usan en ingeniería, física, geología, etc.

##### Potencial eléctrico

Dado un conjunto de $n$ cargas $q_1,\,q_2,\,\cdots,\,q_n$ localizadas en los puntos $\vec{r}_i=x_i\,\mathbf{i}+y_i\,\mathbf{j}+z_i\,\mathbf{k}$ el potencial electrico en el punto $\vec{r}=x\,\mathbf{i}+y\,\mathbf{j}+z\,\mathbf{k}$ está dado por:
$$
V(x,y,z)=\sum_{i=1}^{n}\;\frac{q_i}{\sqrt{(x-x_i)^2+(y-y_i)^2+(z-z_i)^2}}
$$
En esta expresión se ha considerado $K_E=1$ por simplicidad 

- Dipolo eléctrico: $q_1=-q_2=1$, $\vec{r}_1=-\,\mathbf{i}-\,\mathbf{j}$ y $\vec{r}_2=\,\mathbf{i}+\,\mathbf{j}$,
$$
V(x,y,z)=\frac{1}{\sqrt{(x+1)^2+(y+1)^2+z^2}}-\frac{1}{\sqrt{(x-1)^2+(y-1)^2+z^2}}
$$

- Cuatro cargas: las anteriores más $q_3=1$, $q_4=-1$,  $\vec{r}_3=\,\mathbf{k}$ y $\vec{r}_4=-\,\mathbf{k}$
$$
V(x,y,z)=-\frac{1}{\sqrt{x^2+y^2+(z+1)^2}}+\frac{1}{\sqrt{x^2+y^2+(z-1)^2}}-\frac{1}{\sqrt{(x-1)^
   2+(y-1)^2+z^2}}+\frac{1}{\sqrt{(x+1)^2+(y+1)^2+z^2}}
$$

- Cinco cargas: las anteriores mas $q_5=1$, y  $\vec{r}_5=\mathbf{j}$ 
$$
V(x,y,z)=-\frac{1}{\sqrt{x^2+y^2+(z+1)^2}}+\frac{1}{\sqrt{x^2+y^2+(z-1)^2}}-\frac{1}{\sqrt{(x-1)^
   2+(y-1)^2+z^2}}+\frac{1}{\sqrt{(x+1)^2+(y+1)^2+z^2}}
$$

### Campo vectorial

En general uncampo vectorial que toma como entrada un punto $P(x,y,z)$ y devuelve un vector como salida:
\begin{align}
\vec{E}:\mathbb{R}^3\to \mathbb{R}^3
\end{align}

En general un campo vectorial en el espacio tiene la forma:
$$
\vec{E}(x,y,z)=u(x,y,z)\mathbf{i}+v(x,y,z)\mathbf{j}+w(x,y,z)\mathbf{k}
$$
Donde $u,\,v,\,w$ son campos escalares de $\mathbb{R}^3$ en $\mathbb{R}$
Es decir que son funciones de $\mathbb{R}^3$ en $\mathbb{R}^3$:


##### Ejemplo

Ses el campo vectorial:
$$
\vec{E}(x,y,z)=u(x,y,z)\mathbf{i}+v(x,y,z)\mathbf{j}+w(x,y,z)\mathbf{k}
$$

Donde:
\begin{align*}
u(x,y,z) &= \sin(\pi x) \cos(\pi y) \cos(\pi z) \\
v(x,y,z) &= -\cos(\pi x) \sin(\pi y) \cos(\pi z) \\
w(x,y,z) &= \sqrt{\frac{2}{3}} \cos(\pi x) \cos(\pi y) \sin(\pi z)
\end{align*}

Para trazar el campo vecotrial usamos la función `quiver`

```python
import matplotlib.pyplot as plt
import numpy as np

ax = plt.figure().add_subplot(projection='3d')

# Make the grid
x, y, z = np.meshgrid(np.arange(-0.8, 1, 0.2),
                      np.arange(-0.8, 1, 0.2),
                      np.arange(-0.8, 1, 0.8))

# Make the direction data for the arrows
u = np.sin(np.pi * x) * np.cos(np.pi * y) * np.cos(np.pi * z)
v = -np.cos(np.pi * x) * np.sin(np.pi * y) * np.cos(np.pi * z)
w = (np.sqrt(2.0 / 3.0) * np.cos(np.pi * x) * np.cos(np.pi * y) *
     np.sin(np.pi * z))

ax.quiver(x, y, z, u, v, w, length=0.1, normalize=True)

plt.show()
```

##### Ejemplo: Campo inverso del cuadrado de la distancia

El campo gravitacional y el campo eléctrico tienen la forma general:
$$
\vec{E}(x,y,z)=\frac{k}{x^2+y^2+z^2}\left(\frac{x\hat{i}+y\hat{j}+z\hat{k}}{\sqrt{x^2+y^2+z^2}}\right)
$$

Para el caso gravitacional: $k=G\,m_1\,m_2$ donde $G$ es la constante gravitacional y $m_1$, $m_2$ son las masas. En el caso eléctrico: $k=K_e\,q_1\,q_2$ donde $K_e$ es la constante eléctrica y $q_1$, $q_2$ son las cargas. En nuestro caso asumiremos $k=1$ para efectos de ilustrar el comportamiento del campo.

Por lo tanto:

\begin{align*}
u(x,y,z) &= \frac{x}{(x^2+y^2+z^2)^{3/2}}\\
v(x,y,z) &= \frac{y}{(x^2+y^2+z^2)^{3/2}}\\
w(x,y,z) &= \frac{z}{(x^2+y^2+z^2)^{3/2}}
\end{align*}
 Se observa que el campo es radial, es decir que los vectores campo salen desde el origen.

```python
import matplotlib.pyplot as plt
import numpy as np

ax = plt.figure().add_subplot(projection='3d')

# Make the grid
x, y, z = np.meshgrid(np.arange(-0.8, 1, 0.2),
                      np.arange(-0.8, 1, 0.2),
                      np.arange(-0.8, 1, 0.8))

# Make the direction data for the arrows
u = x/(x**2+y**2+z**2)**(3/2)
v = y/(x**2+y**2+z**2)**(3/2)
w = z/(x**2+y**2+z**2)**(3/2)

ax.quiver(x, y, z, u, v, w, length=0.1, normalize=True)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

X, Y = np.meshgrid(np.arange(0, 2 * np.pi, .2), np.arange(0, 2 * np.pi, .2))
U = np.cos(X)
V = np.sin(Y)
fig1, ax1 = plt.subplots()
ax1.set_title(r'Campo vectorial en $2\,D$')
Q = ax1.quiver(X, Y, U, V, units='width')
qk = ax1.quiverkey(Q, 0.9, 0.9, 2, r'$2 \frac{m}{s}$', labelpos='E',
                   coordinates='figure')
plt.show()
```

##### Ejemplo campo vectorial en $2\,D$

$$
\vec{E}(x,y)=(x+y)\mathbf{i}+(x-y)\mathbf{j}
$$

```python
import matplotlib.pyplot as plt
import numpy as np

# Descomenta esta línea para usar un estilo específico de matplotlib sin líneas de cuadrícula
# plt.style.use('_mpl-gallery-nogrid')

# Crea matrices de datos
x = np.linspace(-4, 4, 10)  # Genera números espaciados uniformemente de -4 a 4 con 10 puntos
y = np.linspace(-4, 4, 10)  # Lo mismo para y
X, Y = np.meshgrid(x, y)  # Crea una cuadrícula de puntos 2D a partir de matrices x e y

# Calcula vectores basados en las posiciones de la cuadrícula
U = X + Y  # Componente X del vector en cada punto (suma de x e y)
V = Y - X  # Componente Y del vector (diferencia de y y x)

# Crea la figura y configura los ejes
fig, ax = plt.subplots()

# Dibuja el diagrama de quiver
# - X, Y: definen los puntos de inicio de las flechas
# - U, V: definen los componentes x e y de las direcciones de las flechas
# - color="C0": establece el color de las flechas al primer color del ciclo de color predeterminado
# - angles='xy': especifica que la dirección debe basarse en los valores U y V
# - scale_units='xy': escala la longitud de la flecha según las unidades de datos
# - scale=2: escala la longitud de la flecha por un factor de 2
# - width=.015: establece el ancho de los ejes de las flechas
ax.quiver(X, Y, U, V, color="C0", angles='xy', scale_units='xy', scale=2, width=.01)
ax.set_aspect('equal')
plt.show()
```
