## Vector como $n$-upla de números reales

En 1637, René Descartes extendió la idea de los griegos de usar números para localizar puntos sobre una recta a la idea de utilizar pares de números para localizar puntos en el plano $(a_1, a_2)$ y ternas de números $(a_1, a_2, a_3)$ para localizar puntos en el espacio. 

Posteriormente, A. Cayley (1821-1895) y H. G. Grassmann (1809-1877) se dieron cuenta de que no hay necesidad de limitarse a tres números. Se puede considerar igualmente un cuádruple de números $(a_1, a_2, a_3, a_4)$ o, de manera más general, una $n$-tupla de números reales:
$$
(a_1,a_2,\dots,a_n)\;\;\;\;\;\;n\geq 1
$$
Tales $n$-uplas son llamadas puntos $n$-dimensionales o vectores $n$-dimensionales. El conjunto de números individuales $a_1, a_2, \dots, a_n$ se denomina componentes o coordenadas del vector. La colección de todos los vectores $n$-dimensionales es llamada el espacio vectorial de $n$-uplas.

Al llegar a este punto, surge la pregunta: ¿Por qué nos interesan los espacios de dimensión superior a tres?

Una de las principales razones es que facilitan el análisis de problemas que involucran una gran cantidad de ecuaciones simultáneas. La clave reside en introducir vectores en un espacio $n$-dimensional adecuado, lo que permite reemplazar todas las ecuaciones por una única ecuación vectorial.

Los espacios de dimensión superior a tres no son solo un concepto matemático, sino una herramienta fundamental para comprender el universo y crear nuevas soluciones a los desafíos del mundo actual. Su estudio y aplicación abren un sinfín de posibilidades para el desarrollo científico y tecnológico en el futuro.

La representación de vectores como segmentos de rectas dirigidas (flechas) resulta intuitiva y útil para dimensiones espaciales bajas, como $n=1,2$, y $3$. Sin embargo, esta representación pierde viabilidad cuando la dimensionalidad aumenta ($n>3$). Este hecho conduce a la necesidad de un enfoque completamente analítico para el estudio de vectores en espacios de alta dimensionalidad.

A lo largo de este curso, se empleará una notación consistente para denotar vectores mediante letras romanas mayúsculas en negrita, como $\mathbf{A}$ para referirse al vector 'A'. En ciertas ocasiones, se podrían utilizar letras mayúsculas en itálica, como $\mathit{A}$, o con una flecha superior, como $\vec{A}$. No obstante, se dará preferencia a la tercera forma, ya que resulta fácilmente distinguible, especialmente durante la presentación en pizarrones.

Desde una perspectiva física y geométrica, la identificación de un vector se basa en su magnitud y dirección. La magnitud puede expresarse de dos formas: encapsulando el vector entre barras verticales ($\vert\vert \vec{A}\vert\vert$) o utilizando exclusivamente la letra sin la flecha ($A$). La dirección se caracteriza mediante un vector unitario $\hat{u}_A$ de longitud unitaria ($\vert\vert \hat{u}_A\vert\vert=1$), conocido como vector unitario, que apunta en la misma dirección del vector principal.


Adicionalmente, adoptaremos la convención comúnmente empleada en Python, utilizando corchetes ($[\dots]$) para representar vectores y paréntesis ($(\dots)$) para indicar las coordenadas geométricas de un punto. Es importante señalar que en el contexto de Python, los paréntesis ((...)) denotan típicamente argumentos de funciones y no representan directamente puntos en el espacio. Dependiendo del contexto, los corchetes pueden represntar tanto vectores fila como columna.

Posteriormente, se especificará la notación para representar vectores, utilizando corchetes para vectores y paréntesis para expresar coordenadas geométricas. Se abordará la representación algebraica y geométrica, destacando las diferencias entre vectores y coordenadas.

**Representación algebraica:**

Un vector puede ser descrito como una lista ordenada de números, lo que constituye una interpretación algebraica del mismo. Cada número en la lista representa una coordenada en un sistema de referencia específico. 

**Representación geométrica:**

La representación geométrica de un vector se concibe como un segmento de recta orientado. Este segmento posee una longitud específica, conocida como magnitud, y una dirección, generalmente indicada por un ángulo respecto al eje $x$. La cola del vector señala el punto de origen, mientras que la cabeza indica el punto final, siendo esta última frecuentemente indicada con una flecha.

**Diferencia entre vectores y coordenadas:**

A pesar de las semejanzas iniciales, los vectores y las coordenadas geométricas son conceptos inherentemente diferentes. Las coordenadas representan valores numéricos que posibilitan la ubicación precisa de un punto en el espacio, mientras que un vector no solo indica la posición de un punto, sino también la distancia y la dirección con respecto a un punto de referencia. En otras palabras, un vector posee magnitud, dirección y sentido, mientras que las coordenadas simplemente brindan información acerca de la posición.

Dado lo expuesto, resulta evidente que:
$$
\big[a_1,a_2,\dots,a_n\big]\neq \big(a_1,a_2,\dots,a_n\big)
$$

Desde una perspectiva algebraica, en el ámbito de la física, es común representar un vector por sus componentes en los ejes coordenados $x$, $y$  $z$. Dado un vector $\vec{A}$ en $3\,D$, con componentes $a_x$, $a_y$ y $a_z$, se expresa como:
$$
\vec{A}=a_x\,\mathbf{i}+a_y\,\mathbf{j}+a_z\,\mathbf{k}
$$
Donde $\mathbf{i}$, $\mathbf{j}$ y $\mathbf{k}$ son los vectores unitarios en las direcciones $x$, $y$ y $z$, respectivamente, representando los vectores $\mathbf{i}=\big[1,0,0\big]$, $\mathbf{j}=\big[0,1,\big]$ y $\mathbf{j}=\big[0,1,\big]$.

### Creación de vectores con NumPy

Según el enfoque matemático, un vector se define como una secuencia ordenada de números reales. En el contexto del álgebra lineal abstracta, es importante señalar que los vectores pueden incluir otros objetos matemáticos, como funciones. No obstante, dado que el presente curso se orienta hacia aplicaciones en la Física, se limitará la consideración a vectores conformados exclusivamente por números reales.

En contraste, desde la perspectiva de la codificación, los vectores exhiben características adicionales. Se presentarán las dos primeras a continuación:

- *La dimensionalidad:* Se refiere a la cantidad de números que componen el vector.
- *La orientación:* Varía dependiendo de si el vector presenta una disposición vertical, semejante a una columna, o horizontal, como una fila.

La dimensionalidad se indica como $\mathbb{R}^n$, donde $\mathbb{R}$ indica el conjunto de lo número reales y $n$ indica la dimensionalidad que generalmente se escribe $n\,D$.

Los vectores:
\begin{equation}
\vec{A}=\begin{bmatrix}
1 \\
2 \\
3 \\
4 \\
\end{bmatrix},\hspace{1cm}
\vec{B}=\begin{bmatrix}
1 \\
2 \\
\end{bmatrix},\hspace{1cm}
\vec{C}=\begin{bmatrix} 1 & 2 & 3 & 4 \end{bmatrix}
\end{equation}
$\vec{A}$ es un vector coumna de dimensionalidad $4\,D$, $\vec{B}$ un vector columna de dimensionalidad $2\,D$ y $\vec{C}$ un vector fila de dimensionalidad $4\,D$. Desde el punto de vista matemático, los vectores $\vec{A}$ y $\vec{C}$ son iguales pues tienen las mismas componentes y en el mismo orden. Sin embargo es necesario hacer una precisión desde el punto de vista de la codificación en Pyhton.


Se ha mencionado anteriormente que la dimensionalidad de un vector se define por la cantidad de elementos que contiene. Sin embargo, en el contexto de Python, la dimensionalidad de un vector o matriz se vincula al número de dimensiones geométricas utilizadas para representar un objeto numérico.

Por ejemplo, todos los vectores mostrados anteriormente son considerados "arreglos bidimensionales" en Python, independientemente de la cantidad de elementos que contengan (esto representa la dimensionalidad matemática).

En Python, una lista de números sin una orientación específica se clasifica como un arreglo $1\,D$, sin importar el número de elementos (dicho arreglo se imprimirá como una fila; sin embargo, como se explicará más adelante, se trata de manera diferente a los vectores fila). La dimensionalidad matemática, es decir, el número de elementos en el vector, se denomina longitud o forma(shape) del vector o el arreglo en Python.


En el marco del curso de *Álgebra lineal*, se parte del supuesto de que los vectores adoptan la orientación de vector columna, a menos que se indique lo contrario. Los vectores en formato fila se representan como $\bf{A}^{T}$. Donde $^T$ indica la operación de *transposición*; por el momento, se mencionará que la operación de transposición transforma vectores columna en vectores fila.

En Python, los vectores pueden ser representados mediante diversos tipos de datos. Aunque el tipo *list* (lista) pueda parecer la forma más sencilla de representar un vector, y de hecho lo es en algunas aplicaciones, muchas operaciones de álgebra lineal no funcionarán con listas de Python. Por lo tanto, en la mayoría de los casos, resulta más conveniente crear *vectores como arrays de NumPy*. 

El código siguiente presenta cuatro formas de crear un vector:

```python
#@Diferentes definiciones de un vector en python

# Definición de un vector como lista
asList = [1, 2, 3, 4]  # Lista mutable que contiene 4 elementos

# Definición de un vector como array de NumPy
asArray = np.array([1, 2, 3, 4])  # Array 1D inmutable de NumPy con 4 elementos

# Definición de un vector fila como array de NumPy
filVec = np.array([[1, 2, 3, 4]])  # Array 2D de NumPy con 1 fila y 4 columnas

# Definición de un vector columna como array de NumPy
colVec = np.array([[1], [2], [3], [4]])  # Array 2D de NumPy con 4 filas y 1 columna
#
```

El código define tres variables en Python: asList, asArray, filVec, y colVec.

- `asList` es una lista que contiene los elementos 1, 2, 3 y 4.
- `asArray` es un array unidimensional creado mediante NumPy, compuesto por los mismos elementos 1, 2, 3 y 4.
- `filVec` es un array bidimensional que representa un vector fila, donde todos los elementos están dispuestos en una sola fila.
- `colVec` es un array bidimensional que representa un vector columna, con cada número en una fila separada.

Cabe mencionar que la variable `asArray' se describe como un "*array sin orientación*", indicando que no es estrictamente un vector fila ni un vector columna, sino simplemente una lista unidimensional de números en NumPy. La orientación se especifica utilizando corchetes [..].

En los comentarios del código, se establece que una lista es mutable y un array es inmutable. Dentro del ámbito de la programación, la mutabilidad de un objeto implica la capacidad de modificar su estado interno después de su creación.

Las orientaciones se pueden explorar examinando la forma(shape) de las variables (inspeccionar las formas de las variables a menudo resulta muy útil durante la programación)

```python
#@ Forma(shape) de cada variable
print(f'asList: {np.shape(asList)}')  # Muestra una forma de (4,) como una lista unidimensional
print(f'asArray: {np.shape(asArray)}')# Muestra una forma de (4,) como un array 1D de NumPy
print(f'filVec: {np.shape(filVec)}')  # Muestra una forma de (1, 4) como un array 2D con 1 fila y 4 columnas
print(f'colVec: {np.shape(colVec)}')  # Muestra una forma de (4, 1) como un array 2D con 4 filas y 1 columna
```

La salida muestra que el asArray unidimensional $1\,D$ es de tamaño (4) mientras que los vectores con orientación son arrays bidimensionales y se almacenan con tamaños (1,4) o (4,1) dependiendo de la orientación. Las dimensiones siempre se enumeran como (filas, columnas).

### Estructura de un gráfico en Pyhton


A conticuación se presenta una explicación paso a paso de los elementos clave para generar un gráfico realizado en el código de la celda siguiente:

1. **Importación de Librerías:**
   - La primera línea importa las librerías necesarias, en este caso, `matplotlib.pyplot` como `plt` y `numpy` como `np`. Esta importación es esencial para realizar operaciones de trazado y manipulación de datos.

2. **Definición de Puntos:**
   - A continuación, se definen dos puntos en un espacio bidimensional. `O` representa el origen con coordenadas (0, 0), mientras que `P` representa un punto con coordenadas (3, 4).

3. **Creación de una Figura con Subplot:**
   - Utilizando `plt.subplots()`, se crea una figura y un subplot. `fig` representa la figura completa, y `ax` representa el subplot en el que se trazará la gráfica.

4. **Configuración de los Ejes:**
   - Se establecen los límites de los ejes (`ax.set_xlim(0, 5)` y `ax.set_ylim(0, 5)`) para definir el rango visible en el subplot. 
   - La instrucción `ax.set_aspect('equal')` garantiza que la escala sea igual en ambas direcciones, evitando distorsiones en la visualización.

5. **Configuración de Líneas y Grilla:**
   - Se añaden líneas horizontales y verticales que atraviesan por el origen, logrando una referencia visual. 
   - Además, se configura una grilla punteada (`ax.grid(color='gray', linestyle='--', linewidth=0.5)`) para mejorar la orientación en el subplot.

6. **Scatter Plot del Punto P:**
   - Mediante `ax.scatter()`, se realiza un gráfico de dispersión del punto `P`. Esto añade un marcador en la posición (3, 4) en el subplot.

7. **Configuración del Título del Subplot:**
   - Se agrega un título al subplot mediante `ax.set_title('Gráfica de un punto')` para proporcionar contexto y claridad sobre el contenido del gráfico.

8. **Mostrar el Gráfico:**
   - Finalmente, `plt.show()` se utiliza para presentar y mostrar la figura gráfica construida. Este comando es esencial para visualizar la gráfica generada.

En el siguiente código se detallan los elemento básicos para trazar una gráfica.


```python
#@ Importar las librerías necesarias si ya se han cargado antes no es necesario.

import matplotlib.pyplot as plt
import numpy as np

# Definir los puntos O (origen) y P
O = np.array([0, 0])
P = np.array([3, 4])

# Crear una figura con un subplot
fig, ax = plt.subplots()

# Configuración de límites de los ejes y aspecto igual en los ejes
ax.set_xlim(0, 5)
ax.set_ylim(0, 5)
ax.set_aspect('equal')

# Configuración de las líneas y grilla en los ejes
ax.axhline(0, color='black', linewidth=0.5)
ax.axvline(0, color='black', linewidth=0.5)
ax.grid(color='gray', linestyle='--', linewidth=0.5)

# Scatter plot del punto P
ax.scatter(P[0], P[1])

# Configuración del título del subplot
ax.set_title('Gráfica de un punto usnado la función scatter')

# Mostrar el gráfico
plt.show()
```

### Múltiples  gráficos en Pyhton

La función `plt.subplots(m, n)` de la biblioteca `matplotlib.pyplot` permite crear una cuadrícula de subplots en una figura, donde $m$ representa el número de filas y $n$ el número de columnas. La función devuelve una figura y un array de objetos Axes, uno por cada subplot. Las opciones `sharex=True`y `sharey=True`. 

El siguiente código muestra cómo generar una cuadrícula de $2\times2$ subplots. A cada subplot se le asignan propiedades específicas para sus ejes. Sin embargo, también es posible definir un único conjunto de propiedades para todos los ejes mediante las opciones `sharex` y `sharey`.

```python
#@ Múltiples plots
import matplotlib.pyplot as plt
import numpy as np

#Puntos
O = np.array([0, 0])
P = np.array([3, 4])
Q = np.array([4, 3])

# Crear una figura con tres subplots en 1 fila y 3 columnas
fig, axs = plt.subplots(2, 2, figsize=(5, 5))

# Subplot 11: Gráfica con scatter
axs[0,0].set_xlim(0, 5)
axs[0,0].set_ylim(0, 5)
axs[0,0].set_aspect('equal')

axs[0,0].scatter(P[0], P[1])
axs[0,0].axhline(0, color='black', linewidth=0.5)
axs[0,0].axvline(0, color='black', linewidth=0.5)
axs[0,0].grid(color='gray', linestyle='--', linewidth=0.5)
axs[0,0].set_title('Gráfica de un punto con scatter')

# Subplot 12: Gráfica del vector OP con quiver
axs[0,1].set_xlim(0, 5)
axs[0,1].set_ylim(0, 5)
axs[0,1].set_aspect('equal')

axs[0,1].quiver(O[0], O[1], P[0], P[1], angles='xy', scale_units='xy', scale=1, color='red')
axs[0,1].axhline(0, color='black', linewidth=0.5)
axs[0,1].axvline(0, color='black', linewidth=0.5)
axs[0,1].grid(color='gray', linestyle='--', linewidth=0.5)
axs[0,1].set_title('Vector OP con quiver')

# Subplot 21: Gráfica de la recta OP con plot
axs[1,0].set_xlim(0, 5)
axs[1,0].set_ylim(0, 5)
axs[1,0].set_aspect('equal')

axs[1,0].plot([O[0], P[0]], [O[1], P[1]], color='blue')

axs[1,0].axhline(0, color='black', linewidth=0.5)
axs[1,0].axvline(0, color='black', linewidth=0.5)
axs[1,0].grid(color='gray', linestyle='--', linewidth=0.5)
axs[1,0].set_title('Recta OP con plot')

# Subplot 22: Gráfica de la recta OP con plot
axs[1,1].set_xlim(0, 5)
axs[1,1].set_ylim(0, 5)
axs[1,1].set_aspect('equal')

axs[1,1].plot([P[0], Q[0]], [P[1], Q[1]], color='blue')

axs[1,1].axhline(0, color='black', linewidth=0.5)
axs[1,1].axvline(0, color='black', linewidth=0.5)
axs[1,1].grid(color='gray', linestyle='--', linewidth=0.5)
axs[1,1].set_title('Línea $PQ$ con plot')

# Ajustar la disposición de los subplots
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

### Método de visualización de vectores mediante NumPy y Matplotlib en $2\,D$

La visualización gráfica de vectores desempeña un papel esencial en la comprensión y comunicación de conceptos matemáticos y físicos. En este contexto, se presenta un método efectivo para trazar vectores en un sistema de coordenadas bidimensional mediante el uso de las bibliotecas NumPy y Matplotlib en Python.

Dado un vector $\vec{A}=\begin{bmatrix}
x_0 \\
y_0 \\
\end{bmatrix}$ lo representamos gráficamente en el plano cartesiano como una flecha que une el origen con el punto de coordenadas $(x_0,y_0)$

```python
# Importar la librería matplotlib.pyplot
import matplotlib.pyplot as plt

# Crear una figura y un subplot en 2D
fig, ax = plt.subplots()


# Configurar los límites del eje X y Y entre 0 y 5; Ajustar la relación de aspecto para que sea cuadrada 
ax.set_xlim(0, 5)
ax.set_ylim(0, 5)
ax.set_aspect('equal')

# En los valores siguientes se pueden mofificar  las coordenass de la cola y las componentes de cada vector.
# Coordenadas X de la cola de dos vectores
X = [0, 1]

# Coordenadas Y de la cola de dos vectores
Y = [0, 1]

# Desplazamiento en X de cada vector
U = [4, 1]

# Desplazamiento en Y de cada vector
V = [3, 3]

# Se definen las variables `X`, `Y`, `U` y `V` que contienen 
# las coordenadas de la cola y el desplazamiento de dos vectores.

# Dibujar flechas con kwargs para configuraciones adicionales
plt.quiver(X, Y, U, V, color='green', scale=1, scale_units='xy', angles='xy')

# **Comentario:** Se utiliza la función `plt.quiver` para dibujar dos flechas 
# con diferentes configuraciones. Se establece el color verde, la escala a 1, 
# las unidades de la escala como 'xy' y la dirección de las flechas como 'xy'.

# Mostrar la figura
plt.show()
```

## Operaciones básicas con vectores en $2\,D$

Las operaciones elementales aplicadas a vectores representan un conjunto esencial de herramientas que posibilita la manipulación y análisis efectivo de estas entidades matemáticas. Dentro de este conjunto, destacan operaciones fundamentales tales como el producto por un escalar, la suma, la resta y la combinación lineal

### Producto de un vector por un escalar

En el ámbito matemático, se define un escalar como un número que no se encuentra inserto dentro de una matriz. Por lo general, se representan mediante letras minúsculas, tanto latinas $a,b,\dots,m,n$ como griegas ($\alpha,\beta,\dots$). En este contexto, la expresión que denota el producto de un escalar $c$ por un vector $\vec{A}$ se formula como $c\,\vec{A}$.

Es relevante destacar que, al multiplicar un vector por un escalar, el resultado es nuevamente un vector. 

$$
Escalar\;\;Vector\;⟶\;Vector
$$

En la situación en la que $\vec{A}$ es un vector y $c \neq 0$ es un escalar, el vector resultante $c\,\vec{A}$ conserva la misma dirección que $\vec{A}$, manteniendo constante el ángulo con respecto al eje $x$. La magnitud del vector $c\,\vec{A}$ experimenta una variación proporcional a $\vert c \vert $, y cualquier cambio en la dirección está determinado por el signo de $c$.

Se presentan diversas situaciones resultantes de la multiplicación de un escalar $c$ por un vector $\vec{A}$:

* Cuando $0 < c < 1$, el vector experimenta una contracción.
* En el caso de $c > 1$, el vector experimenta una amplificación.
* Para valores de $c < -1$, el vector se estira e invierte su dirección.
* Si $-1 < c < 0$, el vector se contrae e invierte su dirección.
* Cuando $c = -1$, se invierte la dirección del vector.
* En el caso de $c = 0$, se genera el vector nulo.

#### Colinealidad

Dos vectores no nulos $\vec{A}$ y $\vec{B}$ se dice que son colineales si uno es el multiplo escalar del otro. En otras palabras, si existe un escalar $c$ tal que: $\vec{A}=c\,\vec{B}$

- Los vectores colineales forman una línea recta. Esto significa que si se colocan uno detrás del otro, sus puntos finales coinciden o se encuentran en la misma línea.
- Dos vectores no nulos que no son colineales se denominan no colineales. Estos vectores forman un plano. En otras palabras, si se colocan uno detrás del otro, sus puntos finales no coinciden y no se encuentran en la misma línea.

### Suma de vectores en $2\,D$

#### Regla del triangulo

La suma de dos vectores puede ser representada gráficamente mediante la formación de un triángulo. Para ello, se utiliza la regla del triángulo, también conocida como regla cabeza-cola. Esta técnica permite visualizar la suma de dos vectores $\vec{A}$ y $\vec{B}$, resultando en un nuevo vector  denominado vector suma $\vec{S}=\vec{A}+\vec{B}$.

La aplicación de la regla del triángulo consiste en situar la cola del segundo vector en la cabeza del primero. De esta manera, el vector suma se extiende desde la cola del primer vector hasta la cabeza del segundo.

#### Regla del paralelogramo
Esta regla establece que la suma de dos vectores se puede representar gráficamente mediante un paralelogramo. 
Dados dos vectores, $\vec{A}$ y $\vec{B}$, su suma $\vec{S}=\vec{A}+\vec{B}$  se puede encontrar construyendo un paralelogramo con los vectores $\vec{A}$ y $\vec{B}$. La regla establece que el vector resultante $\vec{S}$ es el diagonal del paralelogramo que parte del punto de origen común a los vectores $\vec{A}$ y $\vec{B}$.

#### Suma algebraica
En el contexto de la suma algebraica de vectores, se parte de los vectores en componentes. Posteriormente, se suman algebraicamente los componentes correspondientes de cada vector. El resultado de esta suma es un nuevo vector que representa la suma algebraica de los vectores originales.

En la representación de vectores columna:
\begin{equation}
\vec{A}=\begin{bmatrix}
a_x \\
a_y \\
\end{bmatrix},\hspace{1cm}
\vec{B}=\begin{bmatrix}
b_x \\
b_y \\
\end{bmatrix},\hspace{1cm}
\end{equation}
Entonces 
\begin{equation}
\vec{S}=\begin{bmatrix}
a_x+b_x \\
a_y+b_y \\
\end{bmatrix}
\end{equation}

### Resta  de vectores en $2\,D$


Para representar gráficamente la resta de dos vectores, $\vec{A}$ y $\vec{B}$, denotada como $\vec{D} = \vec{A} - \vec{B}$, se utiliza la regla del triángulo comúnmente empleada para la suma de vectores. Esta operación se expresa como $\vec{D} = \vec{A} + (-\vec{B})$, indicando que se suma el vector $\vec{A}$ con el negativo de $\vec{B}$.


```python
#Representación geométrica de la resta de vectors.

# Definir vectores
O = np.array([0, 0])
A = np.array([6, 2])
B = np.array([2, 6])
D = A - B  # Vector resultante de la resta A - B

# Crear figura y ejes para el gráfico
fig, ax = plt.subplots()

# Establecer límites del eje x e y en el gráfico
ax.set_xlim(0, 10)
ax.set_ylim(-6, 8)

# Dibujar vectores en el gráfico utilizando la función quiver
ax.quiver(O[0], O[1], A[0], A[1], angles="xy", scale_units='xy', scale=1, color="red", label=r'$\vec{A}$')
ax.quiver(A[0], A[1], B[0], B[1], angles="xy", scale_units='xy', scale=1, color="lightblue", label=r'$-\vec{B}$')
ax.quiver(A[0], A[1], -B[0], -B[1], angles="xy", scale_units='xy', scale=1, color="blue", label=r'$-\vec{B}$')
ax.quiver(O[0], O[1], D[0], D[1], angles="xy", scale_units='xy', scale=1, color="green", label=r'$\vec{D}=\vec{A}-\vec{B}$')

# Mostrar la cuadrícula en el gráfico
ax.grid(True)

# Etiquetas de los ejes x e y
ax.set_xlabel('x(m)')
ax.set_ylabel('y(m)')

# Título del gráfico con notación LaTeX para el vector resultante D
ax.set_title(r'$\vec{D}=\vec{A}-\vec{B}$')

# Añadir etiquetas a los extremos de los vectores A, -B y D
ax.annotate(r'$\vec{A}$', (A[0], A[1]), xytext=(10, 0), textcoords='offset points', fontsize=12, color='red')
ax.annotate(r'$-\vec{B}$', (D[0], D[1]), xytext=(10, 0), textcoords='offset points', fontsize=12, color='blue')
ax.annotate(r'$\vec{B}$', (A[0]+B[0]/2, A[1]+B[1]/2), xytext=(10, 0), textcoords='offset points', fontsize=12, color='lightblue')
ax.annotate(r'$\vec{D}$', (D[0]/2, D[1]/2), xytext=(-10, -20), textcoords='offset points', fontsize=12, color='green')

# Ajustar el aspecto para que las unidades tengan el mismo tamaño
ax.set_aspect('equal')


# Mostrar leyenda en la esquina superior derecha
ax.legend(loc='upper left')

# Mostrar el gráfico
plt.show()
```

Desde un enfoque analítico, la expresión de $\vec{D}=\vec{A}-\vec{B}$ es la siguiente:
$$
\vec{A}=\begin{bmatrix}
a_x \\
a_y \\
\end{bmatrix},\hspace{1cm}
\vec{B}=\begin{bmatrix}
b_x \\
b_y \\
\end{bmatrix},\hspace{1cm}

\vec{D} = \begin{bmatrix}a_x - b_x\\ a_y - b_y\\\end{bmatrix}
$$

Observe que el vector $\vec{D}$ es un vector que va desde la cabeza de $\vec{B}$ hasta la cabeza $\vec{A}$. Generalmente se denota $\vec{BA}$
$$
\vec{BA}=\vec{A}-\vec{B}
$$
Y por lo tanto:
$$
\vec{BA}=-\vec{AB}
$$

Una forma interesante de comprobar lo anterior es observar que:
$$
\vec{B}+\vec{AB}=\vec{A}
$$

```python
#@Representación gráfica de la resta de vectores.

# Definir vectores
O = np.array([0, 0])
A = np.array([6, 2])
B = np.array([2, 6])
D = A - B  # Vector resultante de la resta A - B

# Crear figura y ejes para el gráfico
fig, ax = plt.subplots()

# Establecer límites del eje x e y en el gráfico
ax.set_xlim(0, 8)
ax.set_ylim(0, 8)

# Dibujar vectores en el gráfico utilizando la función quiver
ax.quiver(O[0], O[1], A[0], A[1], angles="xy", scale_units='xy', scale=1, color="red", label=r'$\vec{A}$')
ax.quiver(O[0], O[1], B[0], B[1], angles="xy", scale_units='xy', scale=1, color="blue", label=r'$\vec{B}$')
ax.quiver(B[0], B[1], D[0], D[1], angles="xy", scale_units='xy', scale=1, color="green", label=r'$\overrightarrow{BA}=\vec{A}-\vec{B}$')

# Mostrar la cuadrícula en el gráfico
ax.grid(True)

# Etiquetas de los ejes x e y
ax.set_xlabel('x(m)')
ax.set_ylabel('y(m)')

# Título del gráfico con notación LaTeX para el vector resultante BA
ax.set_title(r'$\overrightarrow{BA}=\vec{A}-\vec{B}$')

# Añadir etiquetas a los extremos de los vectores A, B y BA
ax.annotate(r'$\vec{A}$', (A[0], A[1]), xytext=(10, 0), textcoords='offset points', fontsize=12, color='red')
ax.annotate(r'$\vec{B}$', (B[0], B[1]), xytext=(10, 0), textcoords='offset points', fontsize=12, color='blue')
ax.annotate(r'$\overrightarrow{BA}$', (B[0] + D[0]/2, B[1] + D[1]/2), xytext=(0, 20), textcoords='offset points', fontsize=12, color='green')

# Mostrar leyenda en la esquina superior derecha
ax.legend(loc='upper right')

# Ajustar el aspecto para que las unidades tengan el mismo tamaño
ax.set_aspect('equal')

# Mostrar el gráfico
plt.show()
```

### Combinación lineal de vectores en $2\,D$


Cuando se tienen  tres o mas vectores  $\vec{A}$, $\vec{B}$ y $\vec{C}$, junto con tres o mas escalares $a$, $b$, y $c$, una combinación lineal de estos vectores se expresa como:
$$
\vec{R}=a\,\vec{A}+b\,\vec{B}+c\,\vec{C}
$$
Para obtener esta combinación lineal gráficamente, se multiplican cada vector por su respectivo escalar y luego se aplica la regla del triángulo. Esta regla consiste en colocar el primer vector en el origen, seguido de la cola del segundo vector en la cabeza del primero, y así sucesivamente. El vector resultante tiene su cola en el origen y su cabeza en la cabeza del último vector de la combinación.

```python
# @Representación gráfica de combinación lineal de vectores

# Definir vectores y escalares
O = np.array([0, 0])
A = np.array([6, 2])
B = np.array([2, 6])
C = np.array([4, 1])
a = 1/2
b = 2
c = -1.5
alp=0.50 # Opacidad de los vectores originales A, B y C

# Calcular las combinaciones lineales de forma recurrente
p = a * A
q = a * A + b * B
R = a * A + b * B + c * C  # También se puede usar: r = q + c * w

# Crear figura y ejes para el gráfico
fig, ax = plt.subplots(figsize=(6,6))

# Establecer límites del eje x e y en el gráfico
ax.set_xlim(-1, 15)
ax.set_ylim(-1, 15)

# Dibujar los vectores y combinaciones lineales en el gráfico
ax.quiver(O[0], O[1], A[0], A[1], angles="xy", scale_units='xy', scale=1, color="blue",alpha=alp)
ax.quiver(O[0], O[1], p[0], p[1], angles="xy", scale_units='xy', scale=1, color="blue")


ax.quiver(p[0], p[1],  B[0],  B[1], angles="xy", scale_units='xy', scale=1, color="green",alpha=alp)
ax.quiver(p[0], p[1], b * B[0], b * B[1], angles="xy", scale_units='xy', scale=1, color="green")

ax.quiver(q[0], q[1],   C[0],  C[1], angles="xy", scale_units='xy', scale=1, color="orange",alpha=alp)
ax.quiver(q[0], q[1], c * C[0], c * C[1], angles="xy", scale_units='xy', scale=1, color="orange")


ax.quiver(O[0], O[1], R[0], R[1], angles="xy", scale_units='xy', scale=1, color="red")

# Mostrar la cuadrícula en el gráfico
ax.grid(True)

# Etiquetas de los ejes x e y
ax.set_xlabel('x(m)')
ax.set_ylabel('y(m)')

# Título del gráfico con notación LaTeX para la combinación lineal
ax.set_title(r'$\vec{R}=%s\,\vec{A}+%s\,\vec{B}+(%s)\,\vec{C}$'%(a,b,c))

# Añadir etiquetas a los extremos de los vectores y combinaciones lineales
ax.annotate(r'$%s\;\vec{A}$'%(a), (p[0], p[1]), xytext=(0, -20), textcoords='offset points', fontsize=12, color='blue')
ax.annotate(r'$\vec{A}$', (A[0], A[1]), xytext=(10, -10), textcoords='offset points', fontsize=12, color='blue',alpha=alp)

ax.annotate(r'$%s\;\vec{B}$'%(b), (q[0], q[1]), xytext=(10, -10), textcoords='offset points', fontsize=12, color='green')
ax.annotate(r'$\vec{B}$', (p[0]+B[0]/2, p[1]+B[1]/2), xytext=(10, -10), textcoords='offset points', fontsize=12, color='green',alpha=alp)


ax.annotate(r'$%s\;\vec{C}$'%(c), (R[0], R[1]), xytext=(10, 20), textcoords='offset points', fontsize=12, color='orange')
ax.annotate(r'$\vec{C}$', (q[0]+C[0], q[1]+C[1]), xytext=(-10, -20), textcoords='offset points', fontsize=12, color='orange',alpha=alp)

# Etiqueta para el vector resultante
ax.annotate(r'$\vec{R}$', (R[0]/2, R[1]/2), xytext=(-10, 10), textcoords='offset points', fontsize=12, color='red')

# Ajustar el aspecto para que las unidades tengan el mismo tamaño
ax.set_aspect('equal')


# Mostrar el gráfico
plt.show()
```

De forma analítica, si
\begin{equation}

\end{bmatrix},\hspace{1cm}
\vec{B}=\begin{bmatrix}

\end{bmatrix},\hspace{1cm}

## Operaciones básicas con vectores en $3\,D$

El motivo de separar las operaciones en $2\,D$ y $3\,D$ no es que sean estructuralmente diferentes, lo que cambia es que su representación grafica en Pyhton tiene elementos adicionales

### Definir y trazar la gráfica de un vector en $3\,D$

Definir y trazar un vector en $3\,D$ en esencia es igual que en $2\,D$ sin embargo es necesario hacer precisión en unas cuantas particulariddes.

Un vector $\vec{A}$ en  $3\,D$, con componentes $x_0$, $y_0$ y $z_0$
$$
\vec{A}=\begin{bmatrix}
x_0 \\
y_0 \\
z_0
\end{bmatrix}
$$
Desde el origen, el desplazamiento se realiza $x_0$ unidades en la dirección del eje $x$, seguido por $y_0$ unidades en la dirección paralela al eje $y$ y $z_0$ unidades en la dirección paralela al eje $z$.

### Producto de un vector por un escalar en $3\,D$

Sean $\vec{A}$ un vector y $c$ un escalar, si

$$
\vec{A}=\begin{bmatrix}
x_0 \\
y_0 \\
z_0
\end{bmatrix}
$$

El producto $\vec{B}=c\,\vec{A}$ es un vector con componentes:
$$
\vec{A}=\begin{bmatrix}
c\,x_0 \\
c\,y_0 \\
c\,z_0
\end{bmatrix}
$$

se obtiene multiplicando cada componente de $\vec{A}$ por $c$.

```python
#@ Representación gráfica delproducto de un vector por un escalar en 3D

# Crear una figura y un subplot en 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Establecer límites de los ejes
ax.set_xlim([0, 5])
ax.set_ylim([0, 5])
ax.set_zlim([0, 5])

# Etiquetas de los ejes
ax.set_xlabel("x(m)")
ax.set_ylabel("y(m)")
ax.set_zlabel("z(m)\n",rotation=90)

# Definir los vectores A y B en formato de array y el escalar c
O = np.array([0, 0, 0])
A = np.array([2, 3, 4])
c = 0.5
B = c * A
alp=0.5

# Dibujar el vector B en rojo y el vector A en azul en el gráfico 3D
ax.quiver(O[0], O[1], O[2], B[0], B[1], B[2], color="r", arrow_length_ratio=0.1)
ax.quiver(O[0], O[1], O[2], A[0], A[1], A[2], color="b", arrow_length_ratio=0.1,alpha=alp)

# Etiquetas de los vectores en el gráfico
ax.text(A[0], A[1], A[2], r'$\vec{A}$', color='b', fontsize=12, ha='center', va='bottom',alpha=alp)
ax.text(B[0], B[1], B[2], r'$%s\,\vec{A}$' % (c), color='r', fontsize=12, ha='right', va='bottom')

# Configuración de la vista en el gráfico 3D
ax.view_init(10, 10)

# Título del gráfico
plt.title(r"Producto por un escalar")

# Mostrar el gráfico
plt.show()
```

### Combinación lineal de vectores en $3\,D$

La generalización de las operaciones producto por un escalar y suma de vectores en componentes se puede observar en el siguiente ejemplo.
Si
\begin{align}
\vec{A}&=\begin{bmatrix}
a_x \\
a_y \\
a_z\\
\end{bmatrix},&
\vec{B}&=\begin{bmatrix}
b_x \\
b_y \\
b_z\\
\end{bmatrix},&
\vec{C}&=\begin{bmatrix}
c_x \\
c_y \\
c_z
\end{bmatrix}\\
\end{align}
entonces $\vec{R}=a\,\vec{A}+b\,\vec{B}+c\,\vec{C}$ en componentes toma la forma:
$$
\vec{R}=\begin{bmatrix}
a\,a_x+b\,b_x+c\,c_x \\
a\,a_y+b\,b_y+c\,c_y \\
a\,a_z+b\,b_z+c\,c_z
\end{bmatrix}
$$
donde $a,\,b,\,c$ son escalares.

```python
#@ Crear una figura y un subplot en 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Establecer límites de los ejes
ax.set_xlim([0, 5])
ax.set_ylim([0, 5])
ax.set_zlim([0, 5])

# Etiquetas de los ejes
ax.set_xlabel("x(m)")
ax.set_ylabel("y(m)")
ax.set_zlabel("z(m)")

# Definir los vectores A, B, C y los escalares a, b, c en formato de array
O = np.array([0, 0, 0])
A = np.array([4, 2, 3])
B = np.array([2, 2, 4])
C = np.array([1, 1, 3])
a = 2
b = 1/2
c = -2

# Calcular los vectores de la combinación lineal
p = a * A
q = p + b * B
r = q + c * C

# Dibujar los vectores en el gráfico 3D
ax.quiver(O[0], O[1], O[2], p[0], p[1], p[2], color="b", arrow_length_ratio=0.05)
ax.quiver(p[0], p[1], p[2], b * B[0], b * B[1], b * B[2], color="g", arrow_length_ratio=0.05)
ax.quiver(q[0], q[1], q[2], c * C[0], c * C[1], c * C[2], color="orange", arrow_length_ratio=0.075)
ax.quiver(O[0], O[1], O[2], r[0], r[1], r[2], color="red", arrow_length_ratio=0.05)

# Etiquetas de los vectores en el gráfico
ax.text(p[0], p[1], p[2], r'$%s\vec{A}$'%(a), color='b', fontsize=12, ha='right', va='bottom')
ax.text(q[0], q[1], q[2], r'$%s\vec{B}$'%(b), color='g', fontsize=12, ha='right', va='bottom')
ax.text(r[0], r[1], r[2], r'$%s\vec{C}$'%(c), color='orange', fontsize=12, ha='left', va='top')
ax.text(r[0]/2, r[1]/2, r[2]/2, r'$\vec{R}$', color='r', fontsize=12, ha='left', va='bottom')


# Leyenda en la parte superior derecha del gráfico
ax.legend(loc=(1, 0.1), fontsize=10, frameon=True, fancybox=True, framealpha=0.5,
          labels=[r'$\vec{A}=[%s,%s,%s]$' % (A[0], A[1], A[2]),
                  r'$\vec{B}=[%s,%s,%s]$' % (B[0], B[1], B[2]),
                  r'$\vec{C}=[%s,%s,%s]$' % (C[0], C[1], C[2])])

# Configuración de la vista en el gráfico 3D
ax.view_init(20, 10)

# Título del gráfico
plt.title(r" Regla del triángulo: $\vec{R}=%s\vec{A}+%s\vec{B}+(%s)\vec{C}$" % (a, b, c))

# Mostrar el gráfico
plt.show()
```

## Magnitud de un vector y vector unitario

Un vector unitario, también conocido como vector normalizado, es un vector que tiene magnitud igual a uno. Estos vectores son de gran importancia en diversas áreas de las matemáticas y la física, como se describe a continuación: 

1. Simplificación de cálculos: Los vectores unitarios permiten simplificar cálculos vectoriales, ya que al tener magnitud uno, se eliminan las unidades de medida en las operaciones. Esto facilita el análisis y la comparación de diferentes vectores.

2. Representación de dirección: Los vectores unitarios son ideales para representar direcciones sin importar la magnitud. Por ejemplo, en física, el vector unitario que define la dirección del movimiento de una partícula no depende de la velocidad de la misma.

3. Base de espacios vectoriales: Los vectores unitarios pueden formar una base para un espacio vectorial, lo que significa que cualquier vector en ese espacio puede ser expresado como una combinación lineal de los vectores unitarios de la base.

4. Aplicaciones en física: Los vectores unitarios son omnipresentes en física. Algunos ejemplos de su uso incluyen:
    - Cinemática: El vector unitario que define la dirección de la velocidad o la aceleración.

### Magnitud

La magnitud de un vector, también conocida como longitud geométrica o norma, representa la distancia desde la cola hasta la cabeza de un vector. Esta magnitud se calcula mediante la fórmula estándar de la **distancia euclidiana**, que consiste en tomar la raíz cuadrada de la suma de los cuadrados de las componentes individuales del vector. La expresión matemática para la magnitud de un vector se denota utilizando dobles barras verticales alrededor del vector. Consideremos un vector $\vec{A}$ en un espacio $n-$dimensional expresado como:
$$
\vec{A}=\begin{bmatrix}a_1\\ a_2\\ \vdots \\ a_n\end{bmatrix}
$$
La norma del vector, indicada como $\vert \vert \vec{A} \vert \vert$, se define como:
$$
\vert \vert \vec{A}\vert \vert=\sqrt{\sum_{i=1}^n\,a_i^2}
$$
Donde el símbolo de sumatoria $\big(\sum\big)$ denota la operación sumatoria, es decir, $\sum_{i=1}^n,a_i^2 = a_1^2 + a_2^2 + \cdots + a_n^2$. En ocasiones, para simplificar la notación, la norma del vector se representa con la misma letra latina sin la flecha superior, es decir, $A = \vert \vert \vec{A} \vert \vert$.

En el caso bidimensional ($2\,D$) el vector de posición se expresa como:
$$
\vec{r}=\begin{bmatrix}x_0\\ y_0\\\end{bmatrix}\hspace{1cm} 
$$

y la magnitud de este vector ($\vec{r}$) se calcula utilizando la fórmula de la distancia euclidiana en el plano:
$$
r=\sqrt{x_0^2+y_0^2}
$$

En el caso tridimensional ($3\,D$), el vector de posición se representa como:
$$
\vec{r}=\begin{bmatrix}x_0\\ y_0\\z_0\\\end{bmatrix}\hspace{1cm} 
$$
y la magnitud del vector ($\vec{r}$) se determina utilizando la fórmula de la distancia euclidiana en el espacio:
$$
r=\sqrt{x_0^2+y_0^2+z_0^2}
$$
Estas expresiones corresponden a las medidas de la distancia euclidiana en el plano y en el espacio, respectivamente

### Vector unitario

Sea $\vec{A}$ un vector, se define el vector unitario en la dirección del vector $\vec{A}$, como el vector de magnitud uno que posee la misma dirección que $\vec{A}$. Se denota como $\hat{u}_A$. El vector unitario $\hat{u}_A$ obtiene dividiendo el vector $\vec{A}$ entre su magnitud $A$.

La definición del vector unitario se expresa matemáticamente como:
$$
\hat{u}_{\vec{A}}=\frac{\vec{A}}{A}
$$
La ecuación anterior se puede reescribir como:
$$
\vec{A}=A\,\hat{u}_{\vec{A}}
$$
Esta ecuación indica que todo vector $\vec{A}$ puede sser expresado como el producto de una magnitud $A$ por  un vector unitario $\hat{u}_{\vec{A}}$ que apunta en la direccción de $\vec{A}$.

```python
#@ Vector unitario
A=np.array([4,3])
A_mag=np.linalg.norm(A)
uA=A/A_mag
print(A)
print(A_mag)
print(uA)
```

### Vectores unitarios coordenados cartesianos

El sistema de coordenadas cartesianas es un método para ubicar puntos en un plano bidimensional. Se basa en dos ejes perpendiculares entre sí: el eje horizontal ($X$) y el eje vertical ($Y$). Estos ejes se intersecan en un punto llamado origen ($O$).

Las líneas paralelas a los ejes $X$ e $Y$, que se espacian a intervalos regulares, crean una cuadrícula sobre el plano. Esta cuadrícula facilita la ubicación precisa de puntos mediante coordenadas rectangulares. Las coordenadas rectangulares de un punto son dos números que representan la distancia del punto al origen, medida en unidades a lo largo de los ejes $X$ e $Y$. El primer número es la coordenada $X$ y el segundo es la coordenada $Y$.

Se definen los vectores unitarios:
\begin{align}
\mathbf{i}&=\begin{bmatrix}1\\ 0\\0\\\end{bmatrix},\hspace{1cm}
\mathbf{j}=\begin{bmatrix}0\\ 1\\0\\\end{bmatrix},\hspace{1cm}
\mathbf{k}=\begin{bmatrix}0\\ 0\\1\\\end{bmatrix}
\end{align}

Por lo tanto un vector $\vec{r}=\begin{bmatrix}x_0\\ y_0\\z_0\\\end{bmatrix}$ se representa en la base de vectores ($\mathbf{i}, \mathbf{j}, \mathbf{k}$)
\begin{align}
\begin{bmatrix}x_0\\ y_0\\z_0\\\end{bmatrix}&=
x_0\begin{bmatrix}1\\ 0\\0\\\end{bmatrix}+
y_0\begin{bmatrix}0\\ 1\\0\\\end{bmatrix}+
z_0\begin{bmatrix}0\\ 0\\1\\\end{bmatrix}\\[0.5cm]
\vec{r}&=x_0\mathbf{i}+y_0 \mathbf{j}+z_0 \mathbf{k}
\end{align}

## Producto punto o producto escalar

El *producto punto* (también conocido como *producto interno*) es una de las operaciones más fundamentales en el álgebra lineal. Representa el componente esencial en el cual se fundamentan numerosas operaciones y algoritmos, tales como la convolución, la correlación, la transformada de Fourier, la multiplicación de matrices, la extracción lineal de características, el filtrado de señales, entre otros.

Existen diversas maneras de expresar el producto punto entre dos vectores. Se empleará predominantemente la notación común $\mathbf{A}^T\mathbf{B}$, por razones que se clarifican el al presentar la multiplicación de matrices. En otros contextos, es posible observar $\mathbf{A}\cdot\mathbf{B}$ o $\langle\mathbf{A},\mathbf{B}\rangle$

$$
vector\,\cdot\, vector\to escalar
$$

El producto punto se presenta como un valor singular que proporciona información acerca de la relación entre dos vectores. En primer lugar, se abordará la interpretación de este resultado, para luego centrarse en el algoritmo que permite calcular el producto punto.

#### Interpretación geométrica del producto punto

La definición geométrica del producto punto postula que el resultado de multiplicar dos vectores es igual al producto de sus magnitudes, ajustado por el coseno del ángulo que existe entre ellos.

Sean $\vec{A}$ y $\vec{B}$ dos vectores no nulos:
$$
\vec{A}\cdot\vec{B}=A\,B\,\cos(\theta_{AB})
$$
donde $A$ y $B$ son las magnitudes respectivas de los vectores, y $\theta_{AB}$ es el ángulo entre ellos. Es importante señalar que las magnitudes son estrictamente positivas, mientras que el coseno del ángulo puede variar en el intervalo de $-1$ a $+1$. Este hecho indica que el signo del producto punto queda completamente determinado por la relación geométrica existente entre los dos vectores.

Se presentan cinco casos:
- Ángulo agudo: $0<\theta_{AB}<90^\circ$,   $\cos(\theta_{AB})\,>\,0$ $\wedge$ $\vec{A}\cdot\vec{B}\,>\,0$
- Ángulo obtuso: $\theta_{AB}>90^\circ$,   $\cos(\theta_{AB})\,<\,0$ $\wedge$ $\vec{A}\cdot\vec{B}\,<\,0$
- Ángulo recto(ortogonalidad): $\theta_{AB}=90^\circ$,   $\cos(\theta_{AB})\,=\,0$ $\wedge$ $\vec{A}\cdot\vec{B}\,=\,0$
- Colinealidad: $\theta_{AB}=0^\circ$,  $\cos(\theta_{AB})\,=\,1$ $\wedge$ $\vec{A}\cdot\vec{B}\,=A\,B$
- Colinealidad: $\theta_{AB}=180^\circ$, $\cos(\theta_{AB})\,=\,-1$ $\wedge$ $\vec{A}\cdot\vec{B}\,=-A\,B$

Los vectores ortogonales tienen un producto punto igual a cero (esta afirmación es válida en ambas direcciones: cuando el producto punto es cero, entonces los dos vectores son ortogonales). Por lo tanto, las siguientes afirmaciones son equivalentes: dos vectores son ortogonales; dos vectores tienen un producto punto de cero; dos vectores se cruzan formando un ángulo de 90°.

#### Algoritmo para el cálculo del producto punto

Sean $\vec{A}=[a_1,a_2,\dots\,a_n]$ y $\vec{B}=[b_1,b_2,\dots\,b_n]$, el producto punto se define 
$$
\vec{A}\cdot\vec{B}=\sum_{i=1}^n\,a_i\,b_i
$$
Para calcular el producto punto, se multiplican los elementos correspondientes de los dos vectores y luego se suma sobre todos los productos individuales. En otras palabras, es una multiplicación elemento a elemento seguida de una suma. En la Ecuación $\vec{A}$ y $\vec{B}$ son vectores, y $a_i$ indica el $i-$ésimo elemento de $\vec{A}$.

##### Ejemplo

Dados los vectores $\vec{A}=[1,2,3,4]$ y $\vec{B}=[5,6,7,8]$ calcular el producto punto.
\begin{align}
\vec{A}\cdot \vec{B}&=[1,2,3,4]\cdot [5,6,7,8] \\
&=1*5+2*6+3*7+4*8\\
&=5+12+21+32\\
&=
\end{align}

En el cálculo de producto punto se usa la función `np.dot(A, B)` que realiza el producto punto (dot product) entre dos vectores  $\vec{A}$ y $\vec{B}$ utilizando la función `dot` de la biblioteca NumPy (np). 

```python
#@Producto punto de vectores
A=np.array([1,2,3,4])
B=np.array([5,6,7,8])
resultado=np.dot(A,B)
print(resultado)
```

Nota sobre `np.dot()`: La función np.dot() no implementa realmente el producto punto de vectores; implementa la multiplicación de matrices, que es una colección de productos punto. Esto toma más sentido después de aprender sobre las reglas y mecanismos de la multiplicación de matrices en un curso de Álgebra lineal. Si desea explorar esto ahora, puede modificar el código anterior para dotar a ambos vectores con orientaciones (fila versus columna). Descubrirá que la salida es el producto punto solo cuando el primer input es un vector fila y el segundo input es un vector columna.

```python
#@Producto punto de vectores con producto matricial
A=np.array([1,2,3,4])
B=np.array([[5],[6],[7],[8]])
resultado=A@B # @ indica el producto matricial entre un vector fila A y un vector columna B. En resultado es una matriz
print(resultado)
```

¿Qué sucede si se raliza el producto de dos vectores columna?
Al multiplicar dos vectores columna: $\vec{A}=\begin{bmatrix}1\\2\\3\\4\end{bmatrix}$ y $\vec{B}=\begin{bmatrix}5\\6\\7\\8\end{bmatrix}$

```python
#Producto punto entre vectores columna
A=np.array([[1],[2],[3],[4]])
B=np.array([[5],[6],[7],[8]])
resultado =np.dot(A,B)
```

El error en el código está relacionado con la tentativa de realizar el producto punto entre dos matrices de dimensiones distintas. La matriz $\vec{A}$ presenta una forma de (4, 1), mientras que la matriz $\vec{B}$ posee una forma de (4, 1). El producto punto entre estas matrices no está definido debido a que el número de columnas en la matriz $\vec{A}$ no coincide con el número de filas en la matriz $\vec{B}$.

Para efectuar el producto punto entre dos matrices, es crucial asegurarse de que tengan dimensiones compatibles. En este escenario, si se pretende realizar el producto punto, se puede transponer la matriz $\vec{A}$  para que adquiera una forma de (1, 4) y, posteriormente, llevar a cabo el producto punto. A continuación, se proporciona una corrección al código:

```python
A = np.array([[1], [2], [3], [4]])
B = np.array([[5], [6], [7], [8]])

# Transponer la matriz A para que adquiera una forma de (1, 4)
A_tr = A.T

# Realizar el producto punto
resultado1 = np.dot(A_tr, B)

print(resultado1)
resultado2=A_tr@B
print(resultado2)
```

Para llevar a cabo esta operación en términos de notación matricial, se podría expresar como $\vec{A}^T\; \vec{B}$ donde $\vec{A}^T$ representa la transpuesta del vector columna $\vec{A}$. En esta configuración, se transpone el primer vector columna para convertirlo en un vector fila antes de realizar el producto punto con el segundo vector columna.

Es importante destacar que en el contexto de la multiplicación de vectores columna, la transpuesta de un vector columna se convierte en un vector fila, permitiendo así la realización del producto punto adecuadamente.

#### Propiedades del producto punto

Para todos los vectores $\vec{A}$, $\vec{B}$ y $\vec{C}$ y todos lo escalares $c$, se tiene las siguientes propiedades:
1. Conmutativa: $\vec{A}\cdot\vec{B}=\vec{B}\cdot\vec{A}$
2. Distributiva: $\vec{A}\cdot(\vec{B}+\vec{C})=\vec{A}\cdot\vec{B}+\vec{A}\cdot\vec{C}$
3. Homogeneidad: $c(\vec{A}\cdot\vec{B})=(c\vec{B})\cdot\vec{A}=\vec{B}\cdot(c\vec{A})$
4. No negatividad:  $\vec{A}\cdot\vec{A}\geq 0$ si $\vec{A}\neq 0$
5. Nulidad del producto punto:  $\vec{A}\cdot\vec{A}= 0$ si $\vec{A}= 0$
6. Desigualdad de Cauchy-Schwaz: $\big(\vec{A}\cdot\vec{B}\big)^2\leq \big(\vec{A}\cdot\vec{A}\big)\big(\vec{B}\cdot\vec{B}\big)$

#### Producto punto y perpendicularidad

Dos vectores no nulos, representados como 
$$
\begin{align}
\vec{A}&=\begin{bmatrix}a_1\\a_2\\\vdots\\a_n\\\end{bmatrix}, \hspace{1cm}
\vec{B}=\begin{bmatrix}b_1\\b_2\\\vdots\\b_n\\\end{bmatrix}
\end{align}
$$
son considerados perpendiculares u ortogonales entre sí si su producto punto es igual a cero. 
$$
\vec{A}\cdot\vec{B}=0
$$

En el contexto bidimensional, considerando el vector $\vec{A}=x_0\mathbf{i}+y_0\mathbf{j}$, los vectores $\vec{B}=-y_0\mathbf{i}+x_0\mathbf{j}$ y $\vec{C}=y_0\mathbf{i}-x_0\mathbf{j}$ se caracterizan por ser perpendiculares al vector $\vec{A}$. Es decir:
\begin{align}
\vec{A}\cdot\vec{B}&=x_0\;(-y_0)+y_0\;x_0=0\\
\vec{A}\cdot\vec{B}&=x_0\;(y_0)+y_0\;(-x_0)=0\\
\end{align}
En el contexto bidimensional, resulta relativamente sencillo identificar una dirección perpendicular mediante una rotación de 90°, ya sea en sentido positivo o negativo. Sin embargo, en el caso tridimensional, la situación se torna más compleja. Dada una dirección específica, existen infinitas direcciones perpendiculares a esta, constituyendo, de hecho, un plano en el espacio tridimensional. Más adelante, se introduce la operación de producto cruz, que facilita la determinación de una dirección perpendicular dados dos vectores en un espacio n-dimensional.

Dados dos vectores:
$$
\begin{align}
\vec{A}&=\begin{bmatrix}a_1\\a_2\\\vdots\\a_n\\\end{bmatrix}, \hspace{1cm}
\vec{B}=\begin{bmatrix}b_1\\b_2\\\vdots\\b_n\\\end{bmatrix}
\end{align}
$$
El cálculo del ángulo entre estos vectores se determina mediante la utilización de la definición del producto punto, expresado de la siguiente manera:
\begin{align}
\vec{A}\cdot\vec{B}&=A\,B\,\cos\big(\theta_{AB}\big)\\
\cos\big(\theta_{AB}\big)&=\frac{\vec{A}\cdot\vec{B}}{A\,B}\\
\theta_{AB}&=\cos^{-1}\bigg(\frac{\vec{A}\cdot\vec{B}}{A\,B}\bigg)
\end{align}
Es pertinente destacar que la expresión anterior puede reformularse en términos de los vectores unitarios, con la siguiente representación:
\begin{align}
\theta_{AB}&=\cos^{-1}\Big(\hat{u}_{\vec{A}}\cdot\hat{u}_{\vec{B}} \Big)
\end{align}
Se observa que esta formulación resulta aplicable para la determinación del ángulo entre un par de vectores en un espacio de dimensión arbitraria. Aunque en el caso bidimensional puede considerarse un proceso de naturaleza geométrica simple, en dimensiones mayores o iguales a tres, el procedimiento se complica sustancialmente a primera vista.

En el espacio tridimensional, la determinación de la dirección de un vector mediante un único ángulo se torna impracticable. Esta limitación se debe a la complejidad geométrica que introduce la tercera dimensión. 

En virtud de esta limitación, resulta imperativo recurrir a la definición de los cosenos directores. Estos valores angulares, asociados a cada una de las coordenadas del vector, permiten una descripción completa de su dirección en el espacio tridimensional.

La especificación de los cosenos directores de un vector constituye un procedimiento que normaliza y estandariza la expresión de su dirección en un sistema de coordenadas tridimensional. Esta metodología, independiente de la magnitud del vector, facilita la comparación y el análisis de la orientación espacial entre diferentes vectores.

Dado un vector: $\vec{r}=x_0\mathbf{i}+y_0\mathbf{j}+z_0\mathbf{k}$, la definición de los cosenos directores se formula a partir del producto punto, expresándose como sigue:
\begin{align}
\cos(\theta_x)&=\frac{\vec{A}\cdot \mathbf{i}}{r}=\frac{x_0}{\sqrt{x_0^2+y_0^2+z_0^2}}\\
\cos(\theta_y)&=\frac{\vec{A}\cdot \mathbf{j}}{r}=\frac{y_0}{\sqrt{x_0^2+y_0^2+z_0^2}}\\
\cos(\theta_z)&=\frac{\vec{A}\cdot \mathbf{k}}{r}=\frac{z_0}{\sqrt{x_0^2+y_0^2+z_0^2}}
\end{align}
En cada caso, se determinan los ángulos correspondientes: $\theta_x,\,\theta_y,\, \theta_z$.Es relevante señalar que el vector unitario en la dirección de $\vec{r}$ se expresa como:
$$
\hat{u}_r=\cos(\theta_x)\mathbf{i}+\cos(\theta_y)\mathbf{j}+\cos(\theta_z)\mathbf{k}
$$
Es posible verificar la identidad:
$$
\cos^2(\theta_x)+\cos^2(\theta_y)+\cos^2(\theta_z)=1
$$

#### Proyección y componentes

En esta sección, se dará inicio a la exploración de una descomposición simple pero relevante, que implica la separación de un vector en dos vectores distintos. Uno de estos vectores es ortogonal a un vector de referencia, mientras que el otro es paralelo a dicho vector de referencia. La descomposición ortogonal de vectores conduce de manera directa al procedimiento de Gram-Schmidt y a la descomposición QR, la cual se emplea frecuentemente al abordar problemas inversos en el ámbito estadístico.

Considerando dos vectores $\vec{A}=a_x\mathbf{i}+a_y\mathbf{j}+a_z\mathbf{k}$ y $\vec{B}=b_x\mathbf{i}+b_y\mathbf{j}+b_z\mathbf{k}$  se establece la **proyección escalar** del vector $\vec{B}$ en la dirección del vector $\vec{A}$ como $$Proy_{\vec{A}}\left(\vec{A}\right)=\vec{B}\cdot\hat{u}_{A}$$
Desde una perspectiva geométrica, esta proyección puede interpretarse al alinear las colas de los vectores y proyectar la sombra que $\vec{B}$ genera en la dirección de $\vec{A}$. 

Para visualizarlo con mayor claridad, se puede imaginar un rayo de luz perpendicular a $\vec{A}$ que incide en la cabeza de $\vec{A}$. La sombra resultante sobre $\vec{A}$ representa la proyección escalar de $\vec{B}$ en la dirección de $\vec{A}$. En otras palabras, esta proyección captura la componente de $\vec{B}$ que está alineada con la dirección de $\vec{A}$, y se obtiene mediante el producto escalar entre $\vec{B}$ y el vector unitario $\hat{u}_A$. Este enfoque geométrico ofrece una perspectiva intuitiva sobre la proyección escalar en el contexto de la relación espacial entre los vectores.

Si la proyección escalar se interpreta como una distancia, la **proyección vectorial** se define como ese escalar multiplicado por el vector unitario en la dirección $\hat{u}_A$, constituyendo una cantidad vectorial.
$$
\vec{Proy}_{A}\left(\vec{B}\right)=\left(\vec{B}\cdot\hat{u}_{A}\right)\hat{u}_A
$$

La proyección vectorial $\vec{Proy}_{A}(\vec{B})$ tambien se detona como la componente paralela de $\vec{B}$ en la dirección de $\vec{A}$

$$
\vec{B}_{\parallel\,A}=\vec{Proy}_{A}(\vec{B})
$$
Y permite establecer  la componente perpendicular de $\vec{B}$ en la dirección de $\vec{A}$ apartir de la relación:

$$
\vec{B}=\vec{B}_{\parallel\,A}+\vec{B}_{\perp\,A}
$$
De donde se puede deducir que:
$$
\vec{B}_{\perp\,A}=\vec{B}-\vec{B}_{\parallel\,A}
$$

## Producto cruz o producto vectorial

El producto cruz (o producto vectorial) es otra operación esencial en álgebra lineal que, al igual que el producto punto, desempeña un papel fundamental en diversas aplicaciones y algoritmos. Mientras que el producto punto se caracteriza por generar un escalar que representa la proyección de un vector sobre otro, el producto cruz produce un nuevo vector que es perpendicular al plano definido por los dos vectores originales.

La notación comúnmente utilizada para el producto cruz entre dos vectores $\vec{A}$ y $\vec{B}$ es $\vec{A}\times\vec{B}$. Esta operación no solo proporciona información valiosa sobre la geometría y orientación de los vectores involucrados, sino que también es crucial en el ámbito de la física, gráficos por computadora, mecánica del sólido rígido y otras disciplinas.

$$
vector\,\times\, vector\to vector
$$

La interpretación geométrica del producto cruz radica en la creación de un nuevo vector que es perpendicular al plano formado por los vectores originales, y su magnitud es proporcional al área del paralelogramo que estos vectores delimitan.

#### Definición del producto cruz

En términos generales, la magnitud del producto cruz entre los vectores $\vec{A}$ y $\vec{B}$ se define como:
$$
\vert\vert\vec{A}\times\vec{B}\vert\vert=A\,B\,\sin(\theta_{AB})
$$
Donde $A$ y $B$ representan las magnitudes de los vectores $\vec{A}$ y $\vec{B}$ respectivamente, y $\theta_{AB}$ denota el ángulo entre los dos vectores. La dirección del producto cruz es tal que el resultado apunta perpendicularmente tanto al vector $\vec{A}$ como al vector $\vec{B}$, es perpendicular al plano generado por estos dos vectores. Para determinar la dirección usamos la regla de lamano derecha.


La regla de la mano derecha se puede describir de la siguiente manera:

1. Extiende la mano derecha con los dedos hacia afuera y el pulgar apuntando hacia arriba.
2. Orienta el índice en la dirección del primer vector ($\vec{A}$).
3. Orienta el medio en la dirección del segundo vector ($\vec{B}$).
4. La dirección en que apunta el pulgar representa la dirección del producto cruz ($\vec{A}\times\vec{B}$).

A partir de la definición de producto cruz y la regla de la mano derecha es fácil determinar el producto cruz entre los vectores coordenados unitarios: 
\begin{align*}
\mathbf{i}\times\mathbf{j}&=\mathbf{k}&\mathbf{j}\times\mathbf{i}&=-\mathbf{k}\\
\mathbf{j}\times\mathbf{k}&=\mathbf{i}&\mathbf{k}\times\mathbf{j}&=-\mathbf{i}\\
\mathbf{k}\times\mathbf{i}&=\mathbf{j}&\mathbf{i}\times\mathbf{k}&=-\mathbf{j}\\
\mathbf{i}\times\mathbf{i}&=0&\mathbf{j}\times\mathbf{j}&=0&\mathbf{k}\times\mathbf{k}&=0
\end{align*}

Dados los vectores $\vec{A}=a_x\mathbf{i}+a_y\mathbf{j}+a_z\mathbf{k}$ y $\vec{B}=b_x\mathbf{i}+b_y\mathbf{j}+b_z\mathbf{k}$ el producto cruz:

$$
\vec{A} \times \vec{B} = 
\left( a_y\,b_z - a_z\,b_y \right)\mathbf{i} + 
\left( a_z\,b_x - a_x\,b_z \right)\mathbf{j} + 
\left( a_x\,b_y - a_y\,b_x \right)\mathbf{k}
$$

Que se escribe en forma de determinante:
\begin{align}
\vec{A} \times \vec{B} &= \begin{vmatrix}
    \mathbf{i} & \mathbf{j} & \mathbf{k} \\
    a_x & a_y & a_z \\
    b_x & b_y & b_z
\end{vmatrix}\\
&= \mathbf{i} \begin{vmatrix} a_y & a_z \\ b_y & b_z \end{vmatrix} - \mathbf{j} \begin{vmatrix} a_x & a_z \\ b_x & b_z \end{vmatrix} + \mathbf{k} \begin{vmatrix} a_x & a_y \\ b_x & b_y \end{vmatrix} 
\end{align}

Donde el determinante se calcula de la siguiente forma: $\begin{vmatrix} a_y & a_z \\ b_y & b_z \end{vmatrix}=a_y\,b_z-a_z\,b_y$


#### Interpretación geométrica

En un espacio bidimensional o tridimensional, dos vectores no colineales generan un plano. Este plano se define como la superficie que contiene a todos los puntos que son combinaciones lineales de los dos vectores. Sin pérdida de generalidad, se presenta la ilustración de la interpretación geométrica de la magnitud del producto cruz como el área del paralelogramo formado por los vectores en dos dimensiones.

Dado dos vectores **no colineales** $\vec{A}$ y $\vec{B}$ en el plano, se construye un paralelogramo con estos vectores como lados adyacentes. El área del paralelogramo formado por los vectores $\vec{A}$ y $\vec{B}$ se calcula como el producto de la magnitud de uno de los vectores por la altura del paralelogramo con respecto a ese vector. La altura del paralelogramo con respecto al vector $\vec{A}$ es la magnitud del vector  .

```python
# @Representación gráfica de la proyección

#import matplotlib.patches as patches


# Definir vectores y resultados
O = np.array([0, 0])  # Origen del sistema de coordenadas
A = np.array([6, 2])  # Vector A
B = np.array([2, 6])  # Vector B
# Vectores unitarios coordenados
ux = np.array([1, 0])  # Vector unitario en la dirección x
uy = np.array([0, 1])  # Vector unitario en la dirección y

# Determinación de los ángulos
A_mag = np.linalg.norm(A)  # Magnitud del vector A
B_mag = np.linalg.norm(B)  # Magnitud del vector B

theta_A = round(np.degrees(np.arccos(np.dot(A, ux) / A_mag)), 2)  # Ángulo entre el vector A y el eje x
theta_B = round(np.degrees(np.arccos(np.dot(B, ux) / B_mag)), 2)  # Ángulo entre el vector B y el eje x
theta_m=(theta_A+theta_B)/2

#Suma
S=A+B

# Proyección
BparA=(np.dot(B,A/A_mag))*A/A_mag
BperA=B-BparA


# Crear figura y ejes para el gráfico
fig, ax = plt.subplots()

# Establecer límites del eje x e y en el gráfico
ax.set_xlim(0, 9)
ax.set_ylim(0, 9)

# Configurar ejes y etiquetas personalizadas
ax.set_xlabel('x(m)')
ax.set_ylabel('y(m)')

# Título del gráfico
ax.set_title(r'Área=$A\,h$')

# Etiquetas de los vectores A, B
ax.annotate(r'$\vec{A}$', (A[0], A[1]), xytext=(-10, -20), textcoords='offset points', fontsize=12, color='green')
ax.annotate(r'$\vec{B}$', (B[0], B[1]), xytext=(0, 10), textcoords='offset points', fontsize=12, color='blue')
ax.annotate(r'$h=B\,\sin\left(\theta_{AB}\right)$', (BparA[0]+BperA[0]/2, BparA[1]+BperA[1]/2), xytext=(5, 0), textcoords='offset points', fontsize=12, color='orange')


# Vectores representados mediante flechas
ax.quiver(O[0], O[1], A[0], A[1], angles="xy", scale_units='xy', scale=1, color="green")
ax.quiver(O[0], O[1], B[0], B[1], angles="xy", scale_units='xy', scale=1, color="blue")

# Trazar una línea punteada entre A y B
ax.plot([BparA[0], B[0]], [BparA[1], B[1]], linestyle='dashed', color='gray',linewidth=0.75)
ax.plot([A[0], S[0]], [A[1], S[1]], linestyle='dashed', color='gray',linewidth=0.75)
ax.plot([B[0], S[0]], [B[1], S[1]], linestyle='dashed', color='gray',linewidth=0.75)

# Sombrear la región formada por O, A, S y B en gris
ax.fill_between([O[0], A[0], S[0], B[0],O[0]], [O[1], A[1], S[1], B[1],O[1]], color='gray', alpha=0.3)


# Arcos entre el eje x y los vectores A y B
arc_A = patches.Arc((O[0], O[1]), 4, 4, theta1=0, theta2=theta_A, color='green')
arc_B = patches.Arc((O[0], O[1]), 3, 3, theta1=theta_A, theta2=theta_B, color='blue')
ax.annotate(r'$\theta_{AB}$', (1.25*np.cos(theta_m),1.25* np.sin(theta_m)), xytext=(10, 10), textcoords='offset points', fontsize=12, color='blue')


# Añadir arcos al gráfico
ax.add_patch(arc_B)


# Mostrar la cuadrícula en el gráfico
ax.grid(True)

# Ajustar el aspecto para que las unidades tengan el mismo tamaño
ax.set_aspect('equal')

# Leyenda ubicada en la parte inferior derecha
ax.legend(loc="lower right", fontsize=10, frameon=True, fancybox=True, framealpha=1,
          labels=[r'$\vec{A}=%s\mathbf{i}+%s\mathbf{j}$' % (A[0], A[1]),
                  r'$\vec{B}=%s\mathbf{i}+%s\mathbf{j}$' % (B[0], B[1]),
                  r'$h=%s$' % np.round(np.linalg.norm(BperA),2)],
          handlelength=1, handleheight=0.1
          )

# Mostrar el gráfico
plt.show()
```

El área, $\vert\vert\vec{A}\times\vec{B}\vert\vert$, de la región sobreada es igual al producto de la base $A$ por la altura $h=B\,\sin(\theta_{AB})$

### Coordenadas cartesianas en $2\,D$

Las coordenadas cartesianas en el plano se originan a partir de rectas paralelas a los ejes, que constituyen una rejilla de cuadrados. La ubicación de un punto $(x_0, y_0)$ se especifica mediante un desplazamiento a lo largo del eje $x$ y otro a lo largo del eje $y$, lo que puede expresarse vectorialmente como 
$$\vec{r} = x_0\mathbf{i} + y_0\mathbf{j}$$

Al considerar el triángulo rectángulo formado, se puede concluir que la distancia desde el origen hasta el punto $(x_0, y_0)$, denotada como $r$, es igual a $\sqrt{x_0^2 + y_0^2}$. Asimismo, el ángulo $\theta$ respecto al eje $x$ se obtiene mediante $\theta = \tan^{-1}\left(\frac{y_0}{x_0}\right)$. Estas expresiones coinciden con la definición de la distancia o norma de un vector en dos dimensiones $(2\,D)$.

### Coordenadas  polares $2\,D$

En el sistema de coordenadas polares, en lugar de utilizar una cuadrícula de líneas perpendiculares como en las coordenadas cartesianas, el plano se describe utilizando un sistema de líneas radiales y arcos circulares. Este sistema facilita la descripción de ubicaciones y direcciones en un plano bidimensional de una manera diferente.

En coordenadas polares:

* Las líneas radiales son rayos que se originan en un punto central llamado origen o polo. Estas líneas representan la magnitud o distancia desde el origen al punto en cuestión.

* Los arcos circulares son líneas que se extienden desde el origen y representan el ángulo en sentido antihorario desde la dirección positiva del eje x hasta la línea que conecta el origen con el punto en cuestión.

Este sistema es útil para describir ubicaciones en función de una distancia desde el origen y un ángulo en relación con el eje x. Por ejemplo, un punto en coordenadas polares podría expresarse como "$(r, \theta)$", donde "$r$" es la magnitud (distancia desde el origen) y "$\theta$" es el ángulo.

Las coordenadas polares son especialmente útiles para describir fenómenos que tienen una relación natural con una ubicación central y un ángulo, como la descripción de la posición de un objeto en movimiento a lo largo de una trayectoria circular o el estudio de sistemas con simetría radial.

```python
#@Vector en coordenadas polares
import numpy as np
import matplotlib.pyplot as plt

# Coordenadas polares: magnitud y ángulo en radianes
magnitud = 2.0  # Magnitud del vector
angulo_rad = np.pi / 3  # Ángulo en radianes

# Crear una figura polar
fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(111, polar=True)

# Trazar el vector en coordenadas polares
ax.quiver(angulo_rad, magnitud, color='blue', angles='xy', scale_units='xy', scale=1)

# Configurar el ángulo en el que se encuentra el vector
ax.set_rmax(4.0)  # Establecer la longitud máxima del radio
#ax.set_rticks([])  # Eliminar marcas de radio
ax.set_theta_offset(0)  # Colocar el ángulo cero en la lateral izquierda
ax.set_theta_direction(1)  # Dirección del ángulo: sentido antihorario

ax.set_rmax(2.5)
ax.set_rticks([0.5, 1, 1.5, 2])  # Ticks radiales
ax.set_thetagrids(range(0, 360, 30))


#Título
ax.set_title("Vector en coordenadas polares $(5,60^\circ)$")

# Mostrar la grilla
ax.grid(True)
# Mostrar el gráfico
plt.show()
```

#### Vectores unitarios en coordenadas polares

Se definen los vecores unitarios tangencial $\hat{u}_{r}$ y azimutal $\hat{u}_\theta$ que expresado en coordenadas cortesianas
\begin{align}
\hat{u}_{r}&=\cos(\theta)\mathbf{i}+\sin(\theta)\mathbf{i}\\
\hat{u}_{\theta}&=-\sin(\theta)\mathbf{i}+cos(\theta)\mathbf{i}\\
\end{align}
En general un vector en coordenadas polares se representa:
$$
\vec{r}=r\hat{u}_r+\theta\hat{u}_\theta
$$

#### Transformación de coordenadas: Cartesianas a polares

Cuando contamos con las componentes $x_0$, $y_0$ de un vector, obtenemos los la tupla en coordenadas polared $(r,\theta)$ a partir de la transformación:

$$
 x_0\mathbf{i}+y_0\mathbf{j}\to\begin{cases}
r=\sqrt{x_0^2+y_0^2}\\
\theta=\tan^{-1}\left(\cfrac{y_0}{x_0}\right)
\end{cases}
$$

En *python* calculamos la distancia y el ángulo con el siguiente código

```python
#@Cartesianas a polares

rc=np.array([3,4])
r=np.sqrt(rc[0]**2+rc[1]**2)
theta=np.arctan2(rc[1],rc[0]) #observe el orden coordenada x, coordenada y
ang_deg=np.rad2deg(theta)     #TSe transforman el ángulo en radianes a grados.
(r,ang_deg)
```

Es posible definir una función que convierta coordenadas cartesianas a polares.

```python
#@ Función para el cambio de coordenadas cartesianas a polares
def cart2pol(x, y):
    rho = np.sqrt(x**2 + y**2)
    phi = np.rad2deg(np.arctan2(y, x))
    return(rho, phi)
rc=np.array([3,4])
cart2pol(rc[0],rc[1])
```

#### Transformación de coordenadas: Polares a cartesianas
Cuando se tiene la tupla ordenada en coordenadas polares, se llega a las coordenadas cartesianas a partir de las transformaciones.
$$
\big( r,\theta\big)\to\begin{cases}
x_0=r\,\cos\big(\theta\big)\\
y_0=r\,\sin\big(\theta\big)\\
\end{cases}
$$

Definimos directamente una función que realiza la transformación

```python
def pol2cart(rho, phi):
    x = rho * np.cos(np.deg2rad(phi))
    y = rho * np.sin(np.deg2rad(phi))
    return(x, y)

pol2cart(5,53.13)
```

En coordenadas cilíndricas, se presentan tres vectores unitarios fundamentales: el vector azimutal, denotado como $\hat{u}_\theta$, el vector radial, representado como $\hat{u}_r$ y el vector vertical, indicado por $\hat{u}_z$. Cada uno de estos vectores unitarios despliega una dirección específica en el sistema de coordenadas cilíndricas, siendo $\hat{u}_\theta$ asociado a la dirección azimutal, $\hat{u}_r$ con la dirección radial, y $\hat{u}_z$ con la dirección vertical a lo largo del eje del cilindro.

Expresados en relación con los vectores unitarios en coordenadas cartesianas, los vectores unitarios en coordenadas cilíndricas adquieren la siguiente forma:
\begin{align*}
\hat{u}_\theta&= -\sin(\theta)\mathbf{i}+\cos(\theta)\mathbf{j}\\
\hat{u}_r&= \;\;\;cos(\theta)\mathbf{i}+\sin(\theta)\mathbf{j}\\
\hat{u}_k&=\mathbf{k}\\\\
\end{align*}

#### Conversión entre coordenadas cilíndricas y cartesianas

Las coordenadas rectangulares $(x,y,z)$ y las coordenadas cilíndricas $(r,\theta,z)$ de un punto se relacionan de la siguiente manera:
- Ecuaciones para transformas coordenadas cilíndricas en coordenadas cartesianas
\begin{align}
x&= r\,\cos(\theta)\\
y&= r\,\sin(\theta)\\
z&=z
\end{align}
- Ecuaciones para transformas coordenadas cartesianas  en coordenadas cilíndricas
\begin{align}
r^2&= x^2+y^2\\
\tan(\theta)&= \frac{y}{x}\\
z&=z
\end{align}

### Coordenadas esféricas en $3\,D$.

Al considerar el sistema de coordenadas esféricas, nuevamente utilizamos una triple ordenada para describir la ubicación de un punto en el espacio. En este caso, la triple ordenada describe una distancia y dos ángulos. Las coordenadas esféricas resultan especialmente útiles para describir la geometría de una esfera, de manera análoga a cómo las coordenadas cilíndricas facilitan la descripción de un cilindro. Las líneas de la cuadrícula en el sistema de coordenadas esféricas se construyen en función de las medidas angulares, siguiendo una estructura similar a las coordenadas polares.

En el sistema de coordenadas esféricas, un punto  $𝑃$ en el espacio está representado por una triple coordenada $(\rho,\theta,\phi)$ donde
- $\rho $(la letra griega rho) es la distancia entre  $𝑃$ y el origen.
- $\theta$ es el mismo ángulo utilizado para describir la ubicación en coordenadas cilíndricas;
- $\phi$ (la letra griega phi) es el ángulo formado por el eje $z$ positivo y el segmento de línea $\overline{OP}$ donde $O$ es el origen  y $0\leq\phi\leq\pi$

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from scipy.spatial.transform import Rotation

# Generar puntos en una esfera
phi, theta = np.mgrid[0.0:2.0*np.pi:100j, 0.0:np.pi/2:50j]
x = np.sin(theta) * np.cos(phi)
y = np.sin(theta) * np.sin(phi)
z = np.cos(theta)

#Coordenadas esféricas del puntos
phi0=np.pi/4
theta0=np.pi/6
rho=1

#Puntos auxiliares
p=np.array([rho*np.sin(phi0)*np.cos(theta0),rho*np.sin(phi0)*np.sin(theta0),rho*np.cos(phi0)])
q = np.array([0, 0, p[2]])
r=np.array([p[0], p[1],0])
s=np.array([p[0],0,0])
t=np.array([0,p[1],0])

# Crear la figura y el eje 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

## Vectores unitarios
escala=1
urho=escala*np.array([np.sin(phi0)*np.cos(theta0),np.sin(phi0)*np.sin(theta0),np.cos(theta0)])
utheta=escala*np.array([-np.sin(theta0),np.cos(theta0),0])
uphi=escala*np.array([np.cos(phi0)*np.cos(theta0),np.cos(phi0)*np.sin(theta0),-np.sin(phi0)])

##Trazar utheta, ur y uphi
ax.quiver3D(p[0],p[1],p[2],utheta[0],utheta[1],utheta[2],color='r',arrow_length_ratio=0.075)
ax.quiver3D(p[0],p[1],p[2],urho[0],urho[1],urho[2],color='r',arrow_length_ratio=0.075)
ax.quiver3D(p[0],p[1],p[2],uphi[0],uphi[1],uphi[2],color='r',arrow_length_ratio=0.075)

# Aumentar el tamaño de la figura
fig.set_size_inches(15, 12)

# Graficar la esfera
ax.plot_surface(x, y, z, color='cyan', alpha=0.25, edgecolors='cyan', linewidth=0.025)

# Vector y punto
ax.quiver3D(O[0],O[1],O[2],p[0],p[1],p[2],color='b',arrow_length_ratio=0.075)
ax.scatter(p[0],p[1],p[2],color='b')

# Arco del ángulo theta0
arc=np.linspace(0,theta0,20)
xarc=0.3*np.cos(arc)
yarc=0.3*np.sin(arc)
ax.plot(xarc, yarc, 0*np.ones(20), color='black', linewidth=1)

# Crear arco del angulo phi
## Arco del ángulo phi0
arcphi=np.linspace(0,phi0,20)
zarcphi=0.3*np.cos(arcphi)
xarcphi=0.3*np.sin(arcphi)

##Crear una matriz de rotación para rotar alrededor del eje Z
rotation_matrix = Rotation.from_euler('z', theta0).as_matrix()

##Aplicar la matriz de rotación a los puntos del arco
points_rotated = np.dot(rotation_matrix, np.array([xarcphi, 0*np.ones(20),zarcphi ]))

##Graficar el arco rotada
ax.plot(points_rotated[0], points_rotated[1], points_rotated[2], color='black')

# Agregar la líneas punteadas
ax.plot([p[0], q[0]], [p[1], q[1]], [p[2], q[2]], color='purple', linestyle='--', linewidth=1,alpha=0.5)
ax.plot([p[0], r[0]], [p[1], r[1]], [p[2], r[2]], color='purple', linestyle='--', linewidth=1,alpha=0.5)
ax.plot([s[0], r[0]], [s[1], r[1]], [s[2], r[2]], color='purple', linestyle='--', linewidth=1,alpha=0.5)
ax.plot([t[0], r[0]], [t[1], r[1]], [t[2], r[2]], color='purple', linestyle='--', linewidth=1,alpha=0.5)
ax.plot([O[0], r[0]], [O[1], r[1]], [O[2], r[2]], color='purple', linestyle='--', linewidth=1,alpha=0.5)

# Ajustar el aspecto de los ejes para hacer la esfera más proporcional
ax.set_box_aspect([np.ptp(coord) for coord in [x, y, z]])

# Cálculo de las coordenadas x e y para la circunferencia
t=np.linspace(0,2*np.pi,100)
x_cir = np.cos(t)
y_cir = np.sin(t)

# Dibujar la circunferencia en z=0 y 
ax.plot(x_cir, y_cir, 0*np.ones(100), color='red', linewidth=1)
ax.plot(np.sin(phi0)*x_cir, np.sin(phi0)*y_cir, np.cos(phi0)*np.ones(100), color='red', linewidth=1,linestyle="--")

# Etiquetas
##Coordenadas
ax.text(p[0]*0.75, p[1]*0.75, p[2]*0.75, r'$\rho$', fontsize=14, color='blue', ha='left', va='bottom')
ax.text(p[0]*0.25, p[1]*0.25, p[2]*0.25, r'$\phi$', fontsize=14, color='blue', ha='left', va='bottom')
ax.text(0.55*np.cos(theta0/2),0.55*np.sin(theta0/2) , 0, r'$\theta$', fontsize=14, color='blue', ha='left', va='bottom')

##Vectores unitarios
ax.text(p[0]+1.1*urho[0], p[1]+1.1*urho[1], p[2]+1.1*urho[2], r'$\hat{u}_\rho$', fontsize=14, color='r', ha='center', va='center')
ax.text(p[0]+1.1*utheta[0], p[1]+1.1*utheta[1], p[2]+1.1*utheta[2], r'$\hat{u}_\theta$', fontsize=14, color='r', ha='center', va='center')
ax.text(p[0]+1.1*uphi[0], p[1]+1.1*uphi[1], p[2]+1.1*uphi[2], r'$\hat{u}_\phi$', fontsize=14, color='r', ha='left', va='bottom')

##Coordenadas esféricas del punto
ax.text(*p*1.2, r'$P(\rho,\theta,\phi)$', fontsize=14, color='black', ha='right', va='top')

# Ajustar los límites de los ejes
ax.set_xlim([-1.25, 1.25])
ax.set_ylim([-1.25,1.25])
ax.set_zlim([0, 1.25])

# Etiquetas, título y ajustes adicionales
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title(r'Coordenadas esféricas en $3\,D$')

# Añadir los ejes coordenados
ax.plot([-1.125, 1.125], [0, 0], [0, 0], color='black', linewidth=1.5 )
ax.plot([0, 0], [-1.125, 1.125], [0, 0], color='black', linewidth=1.5 )
ax.plot([0, 0], [0, 0], [0, 1.], color='black', linewidth=1.5)

# Quitar los ticks de los ejes
ax.set_xticklabels([])
ax.set_yticklabels([])
ax.set_zticklabels([])

# Configuración de la vista en el gráfico 3D
ax.view_init(20, 42)

# Mostrar la gráfica
plt.show()
```

#### Vectores unitarios en coordendas esféricas

En coordenadas esféricas, se presentan tres vectores unitarios fundamentales: el vector azimutal, denotado como $\hat{u}_\theta$, el vector radial, representado como $\hat{u}_r$ y el vector , cenital indicado por $\hat{u}_\phi$. Cada uno de estos vectores unitarios despliega una dirección específica en el sistema de coordenadas esféricas, siendo $\hat{u}_\theta$ asociado a la dirección azimutal, $\hat{u}_r$ con la dirección radial, y $\hat{u}_\phi$ con la dirección cenital.

Expresados en relación con los vectores unitarios en coordenadas cartesianas, los vectores unitarios en coordenadas cilíndricas adquieren la siguiente forma:
\begin{align*}
\hat{u}_r&= -\sin(\phi)\,\sin(\theta)\mathbf{i}+\sin(\phi)\,\cos(\theta)\mathbf{j}+\cos(\phi)\,\mathbf{k}\\
\hat{u}_\theta&=-\sin(\theta)\mathbf{i}+\cos(\theta)\mathbf{j}\\
\hat{u}_k&=-\cos(\phi)\,\cos(\theta)\mathbf{i}+\cos(\phi)\,\sin(\theta)\mathbf{j}-\sin(\phi)\,\mathbf{k}\\
\end{align*}

#### Conversión entre coordenadas esféricas y cartesianas


Las coordenadas rectangulares $(x,y,z)$ y las coordenadas esféricas $(\rho,\theta,\phi)$ de un punto se relacionan de la siguiente manera:
- Ecuaciones para transformas coordenadas esféricas en coordenadas cartesianas
\begin{align}
x&= \rho\,\sin(\phi)\cos(\theta)\\
y&= \rho\,\sin(\phi)\sin(\theta)\\
z&=\rho\,\cos(\phi)
\end{align}
- Ecuaciones para transformas coordenadas cartesianas  en coordenadas esféricas
\begin{align}
\rho^2&= x^2+y^2+z^2\\
\tan(\theta)&= \frac{y}{x}\\
\phi&=\cos^{-1}\left(\frac{z}{\sqrt{x^2+y^2+z^2}}\right)
\end{align}

#### Conversión entre coordenadas cilíndricas y esféricas


Si un punto tiene coordenadas cilindricas $(r,\theta,z)$ y  coordenadas esféricas $(\rho,\theta,\phi)$ estas se relacionan de la siguiente manera:
- Ecuaciones para transformas coordenadas esféricas en coordenadas cilindricas
\begin{align}
r&= \rho\,\sin(\phi)\\
\theta&= \theta\\
z&=\rho\,\cos(\phi)
\end{align}
- Ecuaciones para transformas coordenadas cilindricas  en coordenadas esféricas
\begin{align}
\rho^2&= r^2+y^2+z^2\\
\theta&= \phi\\
\phi&=\cos^{-1}\left(\frac{z}{\sqrt{r^2+z^2}}\right)
\end{align}
