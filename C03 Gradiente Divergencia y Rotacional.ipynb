## Derivada direccional y Gradiente

### Definición de gradiente $\nabla$

En el estudio de campos escalares, resulta de gran utilidad la noción de derivada direccional. Esta herramienta permite analizar la tasa de cambio de un campo escalar a lo largo de una dirección específica. Formalmente, la derivada direccional de un campo escalar $\phi$ en un punto $\vec{r}=x\mathbf{i}+y\mathbf{j}+z\mathbf{k}$ en una dirección $\hat{v}$ se define 

$$
D_v\left(\phi(\vec{r})\right)=\lim_{\Delta\,r\to 0}\frac{\phi(\vec{r}+\Delta\,r\,\hat{v})-\phi(\vec{r})}{\Delta\,r}
$$ 

donde 
- $\vec{r}$ es el punto donde se evalua la derivada.
- $\hat{v}$ es el vector unitario que define la dirección a lo largo de la cual se calcula la derivada.
- $\Delta\,r$ es un escalar que representa la magnitud del desplazamiento a lo largo de la dirección $\hat{v}$. 

En términos intuitivos, la derivada direccional nos indica cómo varía el valor del campo escalar $\phi$ al movernos una pequeña distancia desde el punto  $\vec{r}$ en la dirección
$\hat{v}$

Si se considera la dirección $\mathbf{i}$ se tiene que $\Delta\,r=\Delta\,x$:
$$
D_\mathbf{i}\left(\phi(\vec{r})\right)=\lim_{\Delta\,x\to 0}\frac{\phi(\vec{r}+\Delta\,x\,\mathbf{i})-\phi(\vec{r})}{\Delta\,x}
$$ 

La derivada direccional en la dirección $\mathbf{i}$ coincide con la derivada parcial de $\phi$ con respecto a $x$ denotada  $\frac{\partial\,\phi }{\partial x}$. De manera similar, se pueden definir las derivadas direccionales en las direcciones $\mathbf{j},\, \mathbf{k}$ como $\frac{\partial\,\phi }{\partial y}$ y $\frac{\partial\,\phi }{\partial z}$ respectivamente. 

El operador gradiente,$\nabla$, se define como el operador de diferenciación vectorial en coordenadas cartesianas:
$$
\nabla=\mathbf{i}\,\frac{\partial\, }{\partial x}+\mathbf{j}\,\frac{\partial\, }{\partial y} +\mathbf{k}\,\frac{\partial\, }{\partial z}
$$
Cuando el operador gradiente actúa sobre un campo escalar $\phi$ en un punto, el resultado es un vector  $\nabla\,\phi$ que representa tanto la magnitud como dirección de la máxima derivada direccional en ese punto.

El operador gradiente tiene las siguientes propiedades:

1. El gradiente como campo vectorial: El operador gradiente, denotado como $\nabla$, aplicado a un campo escalar $\phi$, produce un campo vectorial. Este campo vectorial, también conocido como gradiente de $\phi$, se representa como $\nabla\,\phi$.

2. Magnitud del gradiente y tasa de cambio: La magnitud del gradiente de $\phi$ denotado  $\vert\vert\nabla\,\phi\vert\vert$, es igual a la máxima tasa de cambio de $\phi$ por unidad de distancia. En otras palabras, indica la rapidez con la que cambia el valor de $\phi$ en la dirección de mayor cambio.

3. Dirección del gradiente y curvas de nivel: el vector $\nabla\,\phi$  apunta en la dirección de máximo cambio de $\phi$ en ese punto. Además, es perpendicular a la curva o superficie de nivel que pasa por dicho punto. Las curvas de nivel son conjuntos de puntos en el espacio donde $\phi$ tiene un valor constante.

4. Cálculo de la derivada direccional:La derivada direccional de $\phi$ en la dirección de un vector unitario $\hat{v}$ se puede calcular mediante la siguiente ecuación
\begin{equation}
D_\mathbf{v}\left(\phi(\vec{r})\right)=\nabla\,\phi\cdot\hat{v}
\end{equation}
Esta expresión representa la proyección del gradiente en la dirección de $\hat{v}$.
5.  Diferencial de un campo escalar: La diferencial de un campo escalar $\phi(\vec{r})$, denotada: $d\,\phi$ se puede calcula mediante el producto escalar del vector gradiente $\nabla\,phi$ por un vector de desplazamiento infinitesimal $\vec{dr}$.
\begin{equation}
d\,\phi(\vec{r})=\nabla\,\phi\cdot d\vec{r}
\end{equation}
Esta expresión indica cuánto cambia la función $\phi(\vec{r})$ cuando se produce un desplazamiento al punto $\phi(\vec{r}+\Delta\,\vec{r})$,  donde $\Delta\,\vec{r}=dx\,\mathbf{i}+dy\,\mathbf{j}+dz\,\mathbf{k}$ es un vector de desplazamiento con componentes en las direcciones $x,\,y\,z$.

##### Ejemplo. Campo escalar en $2\,D$: paraboloide hiperbólico

El campo escalar $\phi(x,y)=x^2-y^2$ genera una superficie en $\mathbb{R}^3$ conocida como paraboloide hiperbólico. Las curvas de nivel asociadas a este campo escalar adoptan la forma de hipérbolas y se dividen en tres categorías distintas:
- Hipérbolas que se extienden en la dirección del eje$x$: Para $\phi(x,y)=c>0$,  las curvas de nivel siguen la ecuación $x^2-y^2=c$. Dichas hipérbolas presentan vértices en $x=\pm c$.
- Hipérbolas que se extienden en la dirección del eje $y$: Para $\phi(x,y)=-c<0$, las curvas de nivel están definidas por $y^2-x^2=c$. Estas hipérbolas tienen vértices en $y=\pm c$.
- Rectas identidad: Cuando $\phi(x,y)=0$, las curvas de nivel adoptan la forma de rectas y siguen la ecuació$x^2-y^2=0$, resultando en líneas rectas de la forma $y=\pm\,x$

El gradiente de la función es:
$$
\nabla\,\phi=2\,x\,\mathbf{i}-2\,y\,\mathbf{i}
$$

Por el punto $(1,0.5)$ pasa la curva de nivel $\phi=0.75$ y el gradiente en ese punto ese: $\nabla\,\phi=2\,\mathbf{i}-\mathbf{j}$. La derivada direccional del campo escalar en el punto $(1,0.5)$ en la dirección $\hat{v}=0.6\mathbf{i}+0.8\mathbf{j}$ es:
$$
\nabla\,\phi(1,0.5)\cdot\hat{v}=0.4
$$
Si calculamos la derivada en la dirección $\hat{v}=\frac{1}{\sqrt{5}}\mathbf{i}+\frac{2}{\sqrt{5}}\mathbf{j}$
\begin{align*}
\nabla\,\phi(1,0.5)\cdot\hat{v}&=(2\,\mathbf{i}-\mathbf{j})\cdot(\frac{1}{\sqrt{5}}\mathbf{i}+\frac{2}{\sqrt{5}}\mathbf{j})\\
&=0
\end{align*}
En este caso la dirección $\hat{v}$ es perpendicular a la dirección del vector gradiente.


###### Código

Definición de la función $\phi=x^2-y^2$, $\nabla\,\phi=2x\mathbf{i}-2y\mathbf{j}$ y  cálculo de la derivada direccional en las direcciones $\vec{V}=3\mathbf{i}+4\mathbf{j}$ y la dirección perpendicular al gradiente en el punto $\vec{r}=0.5\mathbf{i}+0.1\mathbf{j}$

```python
import numpy as np
def phi(x, y):
    return x**2 - y**2

def gradphi(x, y):
    return (2*x, - 2*y)

#Punto
p=np.array([0.5,0.1])

#Valores de phi y el gradiente en el punto
print(f'phi({p[0]},{p[1]})= {phi(*p)}')
print(f'gradphi({p[0]},{p[1]})={gradphi(*p)}')

#Direccióny vector unitario
V=np.array([3,4])
v=V/np.linalg.norm(V)
print(f'v=({v[0]},{v[1]})')

# derivada direccional
der=np.round(np.dot(gradphi(*p),v),5)
print(f'gradphi.v={der}')

# Dirección perperndicular al gradiente en 2D
U=np.array([-gradphi(*p)[1],gradphi(*p)[0]])
# Vector unitario
u=np.round(U/np.linalg.norm(U),4)

print(f'u=({u[0]},{u[1]})')
#Derivada en la dirección tangente a la curva
der=np.dot(gradphi(*p),u)
print(f'gradphi.u={der}')
```

###### Código

Trazado de las curvas de nivel y elcampo gradiente para el campo escalar $\phi(x,y)=x^2-y^2$ tomando $\phi(x,y)=c$ para $c=\{-1,-0.75,-0.5,0.25,0,0.25,0.5,0.75,1\}$. Se traza el punto $\vec{r}_0=0.5\mathbf{i}+0.5\mathbf{j}$ junto con el vector gradiente y el vector unitario en la dirección $\vec{V}=3\mathbf{i}+4\mathbf{j}$

```python
#@ Curvas de nivel, campo gradiente del campo escalar phi(x,y)=x^2+y^2. Gradiente,vector tangene  y vector v en el punto(x0,y0)
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.widgets import Slider

def phi(x, y):
    return x**2 - y**2

def gradphi(x, y):
    return (2*x, - 2*y)

# Función que se obtiene de despejar y haciendo phi(x,y)=c.
def phiy(x,c):
    return np.sqrt(c + x**2)

# Función que se obtiene de despejar x haciendo phi(x,y)=c.
def phix(y,c):
    return np.sqrt(c + y**2)

# Generando la grilla para el trazado de las curvas de nivel.
X, Y = np.meshgrid(np.linspace(-1.5, 1.5, 256), np.linspace(-1.5, 1.5, 256))
Z = phi(X,Y)

#Definición de las valores de c desde -1 hasta 1 con paso de 0.25
levels2 = np.linspace(-1, 1, 9)


def plot_grad_fieldx(c, xmin, xmax, n,ax,color):  # Added 'ax' parameter
    # Vector field
    x = np.linspace(xmin, xmax, n)
    y_positive = phix(x, c)
    y_negative = -phix(x, c)
    
    u, v = gradphi(x,y_positive)
    ax.quiver(x,y_positive, 0.1*u, 0.1*v, color=color,alpha=0.5)

    u, v = gradphi(x,y_negative)
    ax.quiver(x,y_negative, 0.1*u, 0.1*v, color=color,alpha=0.5)

def plot_grad_fieldy(c, ymin, ymax, n,ax,color):  # Added 'ax' parameter
    # Vector field
    y = np.linspace(ymin, ymax, n)
    x_positive = phix(y, c)
    x_negative = -phix(y, c)
    
    u, v = gradphi(x_positive, y)
    ax.quiver(x_positive, y, 0.1*u, 0.1*v, color=color,alpha=0.5)

    u, v = gradphi(x_negative, y)
    ax.quiver(x_negative, y, 0.1*u, 0.1*v, color=color,alpha=0.5)
    
# plot
fig, ax = plt.subplots(figsize=(5, 5))

plot_grad_fieldy(1,-1,1,10,ax,"green")
plot_grad_fieldy(0.25,-1,1,10,ax,"green")
plot_grad_fieldx(1,-1,1,10,ax,"orange")
plot_grad_fieldx(0.25,-1,1,10,ax,"orange")

cs=ax.contour(X, Y, Z, levels=levels2)
ax.set_aspect('equal') 
ax.clabel(cs, fontsize=8)

# Añadir el punto (xp,yp), el gradiente (xp,yp) y el vector unitario en la dirección de V
xp=0.5
yp=0.1
p=np.array([xp,yp])

#Gradiente
gradx,grady=gradphi(*p)
#Vector unitario perpendicular a la curva en (xp,yp)
U=np.array([-grady,gradx])
u=np.round(U/np.linalg.norm(U),2)

#Dirección de V
V=np.array([3,4])#Vector unitario
v=np.round(V/np.linalg.norm(V),2)

#Trazado de los vectores
ax.scatter(p[0], p[1], color='red', s=20,label=r'$\vec{r_0}=%s\mathbf{i}+(%s)\mathbf{j}$'%(p[0],p[1])) 
ax.quiver(p[0], p[1],gradx,grady , color='red',angles='xy',scale_units="xy",scale=1,label=r'$\nabla\,\phi(\vec{r}_0)=%s\mathbf{i}+(%s)\mathbf{j}$'%(gradx,grady))
ax.quiver(p[0], p[1],u[0],v[1] , color='blue',angles='xy',scale_units="xy",scale=1,label=r'$\hat{u}_t=%s\mathbf{i}+(%s)\mathbf{j}$'%(u[0],u[1]))
ax.quiver(p[0], p[1],v[0],v[1] , color='green',angles='xy',scale_units="xy",scale=1,label=r'$\hat{v}=%s\mathbf{i}+(%s)\mathbf{j}$'%(v[0],v[1]))  

ax.set_title(r'Campo  gradiente $\nabla\,\phi(x,y)=2\,x\mathbf{i}-2\,y\mathbf{i}$')
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left') 
plt.show()
```

- La lineas moradas superior e inferior indican $\phi(x,y)=-1$, los vectores naranja indican el campo gradiente sobre dicha curva. 
- La lineas amarillas izquierda y derecha indican $\phi(x,y)=1$, los verdes indican el campo gradiente sobre dicha curva.
- Observe que $\nabla\,\phi\cdot \hat{u}_t=0$ lo que indica que elcampo escalar no cambia en la dirección del vector unitario tangente $\hat{u}_t$ ala curva de nivel que pasa por ese punto. 
- La derivada direccional $D_{v}(\phi(r_0))=\nabla\,\phi\cdot \hat{v}=0.44$ lo que indica una tasa de cambio de $0.44$ unidades en el campo escalar en la dirección del vector $\hat{v}$

###### Código

Superficie generada por el campo escalar $\phi(x,y)=x^2-y^2$, se trazan las curvas de nivel $\phi(x,y)=c$ y se traza el campo gradiente $\nabla\,\phi(x,y)=2\,x\mathbf{i}-2\,y\mathbf{j}$ sobre las curvas de nivel $\phi(x,y)=\pm1$. Observe que el campo gradiente es perpendicular a las curvas de nivel y no a la superfice.

```python
#@ Superficie, curvas de nivel sobre la superficie, campo gradiente sobre curvas de nivel ph(x,y)=1 y ph(x,y)=-1
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

def phi(x, y):
    return x**2 - y**2

def gradphi(x, y):
    return (2*x, -2*y)

def phiy(x,c):
    return np.sqrt(c + x**2)

def phix(y,c):
    return np.sqrt(c + y**2)

def plot_grad_fieldx(c,color):
    # Vector field
    x = np.linspace(-1, 1, 10)
    y_positive = phiy(x,c)
    y_negative = -phiy(x,c)
    
    u, v = gradphi(x, y_positive)
    ax.quiver(x, y_positive, phi(x, y_positive), 0.1*u, 0.1*v, 0, color=color,alpha=0.5)

    u, v = gradphi(x, y_negative)
    ax.quiver(x, y_negative, phi(x, y_negative), 0.1*u, 0.1*v, 0, color=color,alpha=0.5)

def plot_grad_fieldy(c,color):
    # Vector field
    y = np.linspace(-1, 1, 10)
    x_positive = phix(y,c)
    x_negative = -phix(y,c)
    
    u, v = gradphi(x_positive,y)
    ax.quiver(x_positive, y, phi(x_positive, y), 0.1*u, 0.1*v, 0, color=color,alpha=0.5)

    u, v = gradphi(x_negative, y)
    ax.quiver(x_negative, y, phi(x_negative, y), 0.1*u, 0.1*v, 0, color=color,alpha=0.5)


# Ejemplo de uso con c=1
    
# Define the main surface
X, Y = np.meshgrid(np.linspace(-1.5, 1.5, 256), np.linspace(-1.5, 1.5, 256))
Z = phi(X, Y)
# Create the plot
fig, ax = plt.subplots(figsize=(7, 7), subplot_kw={'projection': '3d'})
# Choose a colormap 
cmap = 'viridis'  # Or any other colormap from matplotlib
# Plot the surface
ax.plot_surface(X, Y, Z, cmap=cmap, color='cyan', alpha=0.5, edgecolors='cyan', linewidth=0.025)
# Contour plot
ax.contour(X, Y, Z, cmap=cm.coolwarm, levels=10,alpha=0.25)

#Plot del campo  gradiente
plot_grad_fieldx(1,'green')
plot_grad_fieldy(1.,'orange')

# Añadir el punto (xp,yp), el gradiente (xp,yp) y el vector unitario en la dirección de V
xp=0.5
yp=0.25
zp=phi(xp,yp)
p=np.array([xp,yp,zp])

#Gradiente
gradx,grady=gradphi(p[0],p[1])
#Vector unitario perpendicular a la curva en (xp,yp)
U=np.array([-grady,gradx,0])
u=np.round(U/np.linalg.norm(U),2)

#Dirección de V
V=np.array([3,4,0])#Vector unitario
v=np.round(V/np.linalg.norm(V),2)

ax.scatter(p[0], p[1],p[2], color='red', s=20,label=r'$\vec{r_0}=%s\mathbf{i}+(%s)\mathbf{j}+%s\mathbf{k}$'%(p[0],p[1],p[2])) 
ax.quiver(p[0], p[1],p[2],gradx,grady,0,color='red',label=r'$\nabla\,\phi(\vec{r}_0)=%s\mathbf{i}+(%s)\mathbf{j}$'%(gradx,grady),arrow_length_ratio=0.1)
ax.quiver(p[0], p[1],p[2],u[0],v[1],0,color='gray',label=r'$\hat{u}_t=%s\mathbf{i}+(%s)\mathbf{j}$'%(u[0],u[1]),arrow_length_ratio=0.1)
ax.quiver(p[0], p[1],p[2],v[0],v[1],v[2],color='blue',label=r'$\hat{v}=%s\mathbf{i}+(%s)\mathbf{j}$'%(v[0],v[1]),arrow_length_ratio=0.1)  
#

ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left') 
ax.view_init(20, 30)
plt.show()
```

Se observa que el campo gradiente es perpendicular a las curvas de nivel, y no a la superficie en sí. En cada curva de nivel, el campo gradiente apunta en la dirección (en el plano $xy$) donde el campo escalar experimenta el máximo cambio, es decir, en la dirección donde aumenta su valor.


##### Ejemplo. Campo escalar en $2\,D$: $\phi(x,y)=x\,e^{-x^2-y^2}$

Para el campo escalar: $\phi(x,y)=x\,e^{-x^2-y^2}$ se tiene que:
\begin{align*}
\frac{\partial\,\phi}{\partial x}&=e^{-x^2-y^2}-2 x^2 e^{-x^2-y^2}\\
\frac{\partial\,\phi}{\partial y}&=-2 x y e^{-x^2-y^2}\\
\nabla\,\phi&=\left(e^{-x^2-y^2}-2 x^2 e^{-x^2-y^2}\right)\mathbf{i}-\left(2 x y e^{-x^2-y^2} \right)\mathbf{j}
\end{align*}
Consideremos el punto:$\vec{r}=0.98382\mathbf{i} -0.256141\mathbf{j}$, se tiene que: $\phi(r)=0.35$. El gradiente de la función es:
$$
\nabla\,\phi=-0.33\mathbf{i}+0.18\mathbf{j}
$$
La derivada direccional del campo escalar en el punto $\vec{r}=0.98382\mathbf{i} -0.256141\mathbf{j}$ en la dirección $\hat{v}=0.6\mathbf{i}+0.8\mathbf{j}$ es:
$$
\nabla\,\phi(0.98382,-0.256141)\cdot\hat{v}=0.06
$$
Para el trazado de las  curvas de nivel se tiene en considera la ecuación $c=x\,e^{-x^2-y^2}$. Si se desea despejar $y$ en términos de $c$ y $x$ se obtiene la siguiente expresión
$$y=\log\left(\frac{x}{c}-x^2\right)$$ 
Sin embargo, despejar $x$ en términos de $y$ y $c$ no es tan sencillo. La ecuación no se puede resolver analíticamente para obtener $x$ de forma explícita.

###### Código: Cálculo del campo escalar, el gradiente y la derivada direccional en un punto.

Se definen el campo escalar $\phi(x,y)=\phi(x,y)=x\,e^{-x^2-y^2}$ y el campo gradiente $\nabla\,\phi=\left(e^{-x^2-y^2}-2 x^2 e^{-x^2-y^2}\right)\mathbf{i}-\left(2 x y e^{-x^2-y^2} \right)\mathbf{j}$, ambos campos se evalúan en el punto: $\vec{r}_0=0.98382\mathbf{i}-0.2561\mathbf{j}$. Se calcula la derivada direccional $D_{v}(\phi(\vec{r}_0))$ donde $\hat{v}$ es el vector unitario en la dirección $\vec{V}=3\mathbf{i}+4\mathbf{j}$. Se calcula la derivada direccional de $D_{u}(\phi(\vec{r}_0))$ donde $\vec{u}$ es el vector unitario tangente a la curva de nivel en $\vec{r}_0$

```python
#@ Definición del campo escalar phi(x,y)=x*exp(-x^2+y^2), el campo vectorial gradiente y se calculan las derivadas direccionales en el punto $(0.98382,-0.256141 )$ en la dirección (3,4) y tangente a la curva de nivel que pasa por el punto.
def phi(x, y):
    return x * np.exp(-(x**2 + y**2))

def gradphi(x, y):
    return (np.exp(-(x**2 + y**2)) - 2 * x**2 * np.exp(-(x**2 + y**2)), 
            - 2*y*x * np.exp(-(x**2 + y**2)))

x0=0.98382
y0=-0.256141
#Punto
p=np.array([x0,y0])

#Valores de phi y el gradiente en el punto
print(f'phi({p[0]},{p[1]})= {phi(*p)}')
print(f'gradphi({p[0]},{p[1]})={gradphi(*p)}')

#Direccióny vector unitario
V=np.array([3,4])
v=V/np.linalg.norm(V)
print(f'v=({v[0]},{v[1]})')

# derivada direccional
der=np.round(np.dot(gradphi(*p),v),5)
print(f'gradphi.v={der}')

# Dirección perperndicular al gradiente en 2D
U=np.array([-gradphi(*p)[1],gradphi(*p)[0]])
# Vector unitario
u=np.round(U/np.linalg.norm(U),4)

print(f'u=({u[0]},{u[1]})')
#Derivada en la dirección tangente a la curva
der=np.dot(gradphi(*p),u)
print(f'gradphi.u={der}')
```

###### Código: Trazado de las curvas de nivel y el campo gradiente sobre curvas de nivel del campo escalar.

### Operaciones con $\nabla$

El operador nabla, denotado por $\nabla$, se define como:
$$\nabla=\mathbf{i}\frac{\partial}{\partial x}+\mathbf{j}\frac{\partial}{\partial y}+\mathbf{k}\frac{\partial}{\partial z}$$ 
Donde $\mathbf{i},\, \mathbf{j},\, \mathbf{k}$ son los vectores unitarios en las direcciones $x,\,y,\,z$ respectivamente. $\frac{\partial}{\partial x},\,\frac{\partial}{\partial y},\,\frac{\partial}{\partial z}$
son las derivadas parciales con respecto a $x,\,y,\,z$ respectivamente. 
 
El operador nabla solo tiene sentido cuando se aplica a un objeto matemático, como un campo escalar $\phi$ o un campo vectorial $\vec{h}=h_x\mathbf{i}+h_y\mathbf{j}+h_z\mathbf{k}$.

1.  **Producto por un escalar**: si $\phi$ es un campo escalar el producto de  nabla por $\phi$ es un vector, se define
\begin{equation}
\nabla\,\phi=\mathbf{i}\frac{\partial\,\phi}{\partial x}+\mathbf{j}\frac{\partial\,\phi}{\partial y}+\mathbf{k}\frac{\partial\,\phi}{\partial z}
\end{equation}
Esta operación se conoce como *el gradiente de $\phi$* y representa la tasa de cambio de $\phi$ en cada dirección del espacio.
2. **Producto punto**: El producto punto ($\cdot$) entre dos vectores es un escalar.
\begin{align}
\nabla\cdot\vec{h}&=
\frac{\partial\,h_x}{\partial x}+\frac{\partial\,h_y}{\partial y}+\frac{\partial\,h_z}{\partial z}
\end{align} 
Esta operación se conoce como *la divergencia de $\vec{h}$* y representa la medida en que el campo vectorial se expande o contrae en un punto dado.
3. **Producto cruz**: El producto cruz de nabla con un campo vectorial $\vec{h}$ se define como:
\begin{align}
\nabla \times \vec{h}
&=   \begin{vmatrix}
\mathbf{i} & \mathbf{j} & \mathbf{k} \\
\frac{\partial}{\partial x} & \frac{\partial}{\partial y} & \frac{\partial}{\partial z} \\
h_x & h_y & h_z
\end{vmatrix} \\
 &=
 \left( \frac{\partial h_z}{\partial y} - \frac{\partial h_y}{\partial z} \right) \mathbf{i} + \left( \frac{\partial h_x}{\partial z} - \frac{\partial h_z}{\partial x} \right) \mathbf{j} + \left( \frac{\partial h_y}{\partial x} - \frac{\partial h_x}{\partial y} \right) \mathbf{k} \\
\end{align}
Esta operación se conoce como *el rotacional de $\vec{h}$* y representa la medida en que el campo vectorial gira en un punto dado.

Utilizando estas tres combinaciones, podemos describir las variaciones espaciales de los campos de una manera conveniente y general, en el sentido de que es independiente del sistema de ejes que se tome.

### Campos vectoriales

En el análisis de los campos vectoriales, se identifican dos propiedades fundamentales que posibilitan una descripción precisa de su comportamiento: el flujo y la circulación. Para una comprensión más detallada de estas propiedades, se puede concebir un campo vectorial de velocidades que caracteriza el movimiento de un líquido en una piscina.

#### **Flujo**
Al analizar un campo vectorial de velocidades que describe el movimiento de un fluido, se define el flujo como la cantidad de fluido que atraviesa una superficie cerrada en un intervalo de tiempo específico. Esta propiedad facilita la determinación de la existencia de un flujo neto, ya sea hacia adentro o hacia afuera de la superficie, así como la evaluación de la magnitud del flujo en diferentes regiones.

Se denomina flujo de velocidad saliente de la superficie a la cantidad de fluido que sale por unidad de tiempo. El flujo a través de un elemento de superficie es precisamente igual al producto de la componente de la velocidad normal a la superficie por el área de dicha superficie. 

Para una superficie cerrada cualquiera, el flujo neto, o simplemente flujo, se obtiene multiplicando el promedio de la componente de la velocidad en dirección hacia afuera por el área de la superficie.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Definir la función phi
def phi(x, y):
    return 1 - x**2 - y**2

# Definir la función gradphi
def gradphi(x, y):
    return (-2*x, -2*y)

# Generar datos para la superficie en coordenadas polares
theta = np.linspace(0,2*np.pi, 100)
r = np.linspace(0, 1, 100)
theta, r = np.meshgrid(theta, r)
x = r * np.cos(theta)
y = r * np.sin(theta)
z = phi(x, y)

# Punto específico (x0, y0, z0)
x0, y0 = 0.5, 0.5
z0 = phi(x0, y0)

# Crear la figura y los ejes 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Trazar la superficie
#ax.plot_surface(x, y, z, cmap='viridis')

##
grad3d=-np.array([gradphi(x0, y0)[0], gradphi(x0, y0)[1],-1])
normal=0.5*grad3d/np.linalg.norm(grad3d)
# Dibujar el vector con quiver
ax.quiver(x0, y0, z0, normal[0], normal[1], normal[2], color='red')
ax.quiver(x0, y0, z0, 0, 0, 0.5, color='blue')

## Plano tangente
# Generar datos para la superficie el plano tangente
thetap = np.linspace(0,2*np.pi, 100)
rp = np.linspace(0, 0.5, 100)
thetap, rp = np.meshgrid(thetap, rp)
xp = x0+normal[0]*rp * np.cos(thetap)
yp = y0+normal[1]*rp * np.sin(thetap)
zp = z0+normal[2]*phi(xp, yp)
ax.plot_surface(xp, yp, zp,color='yellow')

ax.scatter(x0, y0, z0)
# Ejes proporcionales

# Ajustar la escala de los ejes individualmente
ax.set_xlim([min(x.flatten()), max(x.flatten())])
ax.set_ylim([min(y.flatten()), max(y.flatten())])
ax.set_zlim([min(z.flatten()), max(z.flatten())])


# Etiquetas de los ejes
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Configuración de la vista en el gráfico 3D
ax.view_init(20, 80)
# Mostrar el gráfico
#plt.legend()
plt.show()
```
